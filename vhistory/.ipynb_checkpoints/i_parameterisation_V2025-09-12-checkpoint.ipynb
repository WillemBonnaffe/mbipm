```python
##############
## INITIATE ##
##############

## Imports
import torch
import numpy as np
import matplotlib.pyplot as plt
import time
import seaborn as sns

## Figure parameters
FIGSIZE = (8,6)
plt.rcParams['figure.dpi'] = 100
custom_params = {"axes.spines.right": False, 
                 "axes.spines.top": False, 
                 'axes.facecolor':'whitesmoke'}
sns.set_theme(style="ticks", rc=custom_params)

#
###
```

```python
#######################
## WILLOW - ONTOGENY ##
#######################

## Function
def W(H, alpha, b=-0.459, c=0.782):
    return alpha * (b + c * H)

## Parameters
H_infty = 4
alpha = 1.33

## Evaluate
W_infty = W(H_infty, alpha)
print(W_infty)

#
###
```

```python
#######################
## WILLOW - SURVIVAL ##
#######################

## Figure variables
YLABEL = 'Annual Survival Rate (yr$^{-1}$)'
FIGNAME = 'figure-parameterisation-willow-survival.png'

## Function
def y(x, ttheta):
    return ttheta[0]/(1 + np.exp(-(x - ttheta[1]) * ttheta[2]))

## Parameters
ttheta = np.array([0.983, 0.100, 5.000])

## Evaluate function
xx = np.arange(0, 5, 0.01)
yy_ct = y(xx, ttheta)
    
## Visualise
plt.figure(figsize=FIGSIZE)
plt.plot(xx, yy_ct, label='Prior estimate', color='black')
plt.xlabel('Biomass (kg)')
plt.ylabel(YLABEL)
# plt.ylim(0,1)
plt.legend(frameon=False, loc='lower right')
# plt.savefig(FIGNAME)
plt.show()

#
###
```

```python
##############################
## WILLOW - NUTRIENT UPTAKE ##
##############################

## Nitrogen uptake rate of willow
uptake = 0.02 # (Âµmol N / min)
molar_mass_nitrogen = 14.007 # (kg N / mol)
uptake = molar_mass_nitrogen * uptake * 10**(-6) * 60 * 24 * 365 * 0.5 # (kg N / year)
print(uptake)
mass_total_exp = 17 * 1e-3 # (kg willow)
uptake_nitrogen_willow = uptake / mass_total_exp # (kg N / kg willow / year)
print(uptake_nitrogen_willow)
print('')

## Total biomass consumption
nitrogen_efficiency = (148 + 169 + 148 + 108)/4 # (na) (Weih et al. 2021, p. 4(4), Table 4)
print(nitrogen_efficiency)
yearly_biomass_consumption_willow = uptake_nitrogen_willow * nitrogen_efficiency
print(yearly_biomass_consumption_willow)
print('')

"""
## Nitrogen uptake rate of willow (alternative)
uptake = 0.02 # (mg N / h) 
uptake = uptake * 1e-6 * 24 * 365 * 0.5 # (kg N / year)
mass_total_exp = 17 * 1e-3 # (kg willow)
uptake_nitrogen_willow = uptake / mass_total_exp # (kg N / kg willow / year)
print(uptake_nitrogen_willow)
print('')

## Total biomass consumption
nitrogen_efficiency = (148 + 169 + 148 + 108)/4 # (na) (Weih et al. 2021, p. 4(4), Table 4)
yearly_biomass_consumption_willow = uptake_nitrogen_willow * nitrogen_efficiency
print(yearly_biomass_consumption_willow)
print('')
"""

## Nitrogen uptake relative to availability
available_nitrogen = 40 * 1e-6 * 1e4 * 365 # (kg / m^2 / year) Geremia et al. (2025), Fig. 3, C
print(available_nitrogen) # (kg / m^2 / year)
print(uptake_nitrogen_willow/available_nitrogen)
print('')

#
###
```

```python
##########################
## WILLOW - CONSUMPTION ##
##########################

## Figure variables
YLABEL = 'Asymptotic Consumption Rate (kg N.yr$^{-1}$)'
FIGNAME = 'figure-parameterisation-willow-consumption-asymptotic.png'

## Function
def y(x, ttheta):
    return ttheta[0] * 0.001 / (ttheta[1] * ttheta[2]) * 24 * 365 * x

## Parameters
ttheta_ct = np.array([0.02, 1.0, 0.01])
ttheta_lb = np.array([0.015, 1.0, 0.005])
ttheta_hb = np.array([0.025, 1.0, 0.02])

## Evaluate xâaxis
xx = np.arange(0, 5, 0.01)

## Evaluate CT
yy_ct = y(xx, ttheta_ct)

## Sample N random parameter sets from the [low,high] ranges
num_samps = 10000
tttheta_samps = []    
for i in range(ttheta_ct.shape[0]): 
    ttheta_samps = np.random.uniform(ttheta_lb[i], ttheta_hb[i], num_samps)
    tttheta_samps += [ttheta_samps]
tttheta_samps = np.stack(tttheta_samps)
print(tttheta_samps.shape)

## Compute median and CI
CI = np.percentile([y(1, tttheta_samps[:,k]) for k in range(num_samps)], [2.5, 97.5])
midpoint = np.percentile([y(1, tttheta_samps[:,k]) for k in range(num_samps)], [50])
print(CI)
print(midpoint)

## Compute the sampleâcurve values
yy_samps = [y(xx, tttheta_samps[:,k]) for k in range(num_samps)]

## Compute percentiles
ci95_lower, ci95_upper = np.percentile(yy_samps, [2.5, 97.5], axis=0)
ci50_lower, ci50_upper = np.percentile(yy_samps, [25, 75],   axis=0)
yy_med = np.percentile(yy_samps, 50, axis=0)
    
## Visualise
plt.figure(figsize=FIGSIZE)
# for yy in yy_samps:
#    plt.plot(xx, yy, color='grey', alpha=0.2)
plt.plot(xx, yy_ct, label='Midpoint', color='salmon')
plt.plot(xx, yy_med, label='Median', color='grey')
# plt.plot(xx, xx, '--', color='black', lw=1.0)
plt.fill_between(xx, ci95_lower, ci95_upper, color='grey', alpha=0.2, label='95% CI')
plt.fill_between(xx, ci50_lower, ci50_upper, color='grey', alpha=0.4, label='50% CI')
plt.xlabel('Biomass (kg)')
plt.ylabel(YLABEL)
# plt.ylim(0,1)
plt.legend(frameon=False, loc='lower right')
plt.savefig(FIGNAME)
plt.show()

#
###
```

```python
###########################
## WILLOW - REPRODUCTION ##
###########################

## Figure variables
YLABEL = 'Probability of Reproduction'
FIGNAME = 'figure-parameterisation-willow-survival.png'

## Function
def y(x, ttheta):
    return ttheta[0]/(1 + np.exp(-(x - ttheta[1]) * ttheta[2]))

## Parameters
ttheta = np.array([0.4, 0.75, 10.000])

## Evaluate function
xx = np.arange(0, 5, 0.01)
yy_ct = y(xx, ttheta)
    
## Visualise
plt.figure(figsize=FIGSIZE)
plt.plot(xx, yy_ct, label='Prior estimate', color='black')
plt.xlabel('Biomass (kg)')
plt.ylabel(YLABEL)
# plt.ylim(0,1)
plt.legend(frameon=False, loc='lower right')
# plt.savefig(FIGNAME)
plt.show()

#
###
```
