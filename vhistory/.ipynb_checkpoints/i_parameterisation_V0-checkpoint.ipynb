```python
##############
## INITIATE ##
##############

## Imports
import torch
import numpy as np
import matplotlib.pyplot as plt
import time
import seaborn as sns

## Figure parameters
FIGSIZE = (8,6)
plt.rcParams['figure.dpi'] = 100
custom_params = {"axes.spines.right": False, 
                 "axes.spines.top": False, 
                 'axes.facecolor':'whitesmoke'}
sns.set_theme(style="ticks", rc=custom_params)

#
###
```

```python
#####################
## WILLOW ONTOGENY ##
#####################

def W(H, alpha, b=-0.459, c=0.782):
    return alpha * (b + c * H)

## Parameters
H = [4, 3, 5]
alpha = [1.33, 1.23, 1.42]

## Evaluate
W_m = W(H[0], alpha[0])
W_lo = W(H[1], alpha[1])
W_hi = W(H[2], alpha[2])

## Print
print(W_m)
print(W_lo)
print(W_hi)

#
###
```

```python
#####################
## WILLOW SURVIVAL ##
#####################

## Figure variables
YLABEL = 'Annual Survival Rate (yr$^{-1}$)'
FIGNAME = 'figure-parameterisation-willow-survival.png'

## Function
def y(x, ttheta):
    return ttheta[0] + (ttheta[1]-ttheta[0])/(1 + np.exp(-(x - ttheta[2]) * ttheta[3]))

## Parameters
ttheta_ct = np.array([0.540, 0.983, 0.400, 5.000])
ttheta_lb = np.array([0.130, 0.975, 0.200, 5.000])
ttheta_hb = np.array([0.960, 0.987, 1.500, 5.000])

## Evaluate xaxis
xx = np.arange(0, 5, 0.01)

## Evaluate CT
yy_ct = y(xx, ttheta_ct)

## Sample N random parameter sets from the [low,high] ranges
num_samps = 100
tttheta_samps = []    
for i in range(ttheta_ct.shape[0]): 
    ttheta_samps = np.random.uniform(ttheta_lb[i], ttheta_hb[i], num_samps)
    tttheta_samps += [ttheta_samps]
tttheta_samps = np.stack(tttheta_samps)
print(tttheta_samps.shape)

## Compute the samplecurve values
yy_samps = [y(xx, tttheta_samps[:,k]) for k in range(num_samps)]

## Compute percentiles
ci95_lower, ci95_upper = np.percentile(yy_samps, [2.5, 97.5], axis=0)
ci50_lower, ci50_upper = np.percentile(yy_samps, [25, 75],   axis=0)
yy_med = np.percentile(yy_samps, 50, axis=0)
    
## Visualise
plt.figure(figsize=FIGSIZE)
# for yy in yy_samps:
#    plt.plot(xx, yy, color='grey', alpha=0.2)
plt.plot(xx, yy_ct, label='Midpoint', color='salmon')
plt.plot(xx, yy_med, label='Median', color='grey')
# plt.plot(xx, xx, '--', color='black', lw=1.0)
plt.fill_between(xx, ci95_lower, ci95_upper, color='grey', alpha=0.2, label='95% CI')
plt.fill_between(xx, ci50_lower, ci50_upper, color='grey', alpha=0.4, label='50% CI')
plt.xlabel('Biomass (kg)')
plt.ylabel(YLABEL)
# plt.ylim(0,1)
plt.legend(frameon=False, loc='lower right')
plt.savefig(FIGNAME)
plt.show()

#
###
```

```python
########################
## WILLOW CONSUMPTION ##
########################

## Figure variables
YLABEL = 'Asymptotic Consumption Rate (kg N.yr$^{-1}$)'
FIGNAME = 'figure-parameterisation-willow-consumption-asymptotic.png'

## Function
def y(x, ttheta):
    return ttheta[0] * 0.001 / (ttheta[1] * ttheta[2]) * 24 * 365 * x

## Parameters
ttheta_ct = np.array([0.02, 1.0, 0.01])
ttheta_lb = np.array([0.015, 1.0, 0.005])
ttheta_hb = np.array([0.025, 1.0, 0.02])

## Evaluate xaxis
xx = np.arange(0, 5, 0.01)

## Evaluate CT
yy_ct = y(xx, ttheta_ct)

## Sample N random parameter sets from the [low,high] ranges
num_samps = 10000
tttheta_samps = []    
for i in range(ttheta_ct.shape[0]): 
    ttheta_samps = np.random.uniform(ttheta_lb[i], ttheta_hb[i], num_samps)
    tttheta_samps += [ttheta_samps]
tttheta_samps = np.stack(tttheta_samps)
print(tttheta_samps.shape)

## Compute median and CI
CI = np.percentile([y(1, tttheta_samps[:,k]) for k in range(num_samps)], [2.5, 97.5])
midpoint = np.percentile([y(1, tttheta_samps[:,k]) for k in range(num_samps)], [50])
print(CI)
print(midpoint)

## Compute the samplecurve values
yy_samps = [y(xx, tttheta_samps[:,k]) for k in range(num_samps)]

## Compute percentiles
ci95_lower, ci95_upper = np.percentile(yy_samps, [2.5, 97.5], axis=0)
ci50_lower, ci50_upper = np.percentile(yy_samps, [25, 75],   axis=0)
yy_med = np.percentile(yy_samps, 50, axis=0)
    
## Visualise
plt.figure(figsize=FIGSIZE)
# for yy in yy_samps:
#    plt.plot(xx, yy, color='grey', alpha=0.2)
plt.plot(xx, yy_ct, label='Midpoint', color='salmon')
plt.plot(xx, yy_med, label='Median', color='grey')
# plt.plot(xx, xx, '--', color='black', lw=1.0)
plt.fill_between(xx, ci95_lower, ci95_upper, color='grey', alpha=0.2, label='95% CI')
plt.fill_between(xx, ci50_lower, ci50_upper, color='grey', alpha=0.4, label='50% CI')
plt.xlabel('Biomass (kg)')
plt.ylabel(YLABEL)
# plt.ylim(0,1)
plt.legend(frameon=False, loc='lower right')
plt.savefig(FIGNAME)
plt.show()

#
###
```

```python
########################
## WILLOW CONSUMPTION ##
########################

## State variables
R = 1000
N = 100
Z = 5
XX = [R, N, Z]

## Function
def y(xx, ttheta):
    return ttheta[0] * xx[0] * ttheta[1] * xx[1] * xx[2] / (ttheta[0] * xx[0] + ttheta[1] * xx[1] * xx[2])
    # return ttheta[0] * xx[0] * ttheta[1] * xx[1] * xx[2] / (ttheta[0] * xx[0] + ttheta[1] * xx[1] * xx[2])

## Parameters
ttheta_ct = np.array([1.0, 1.03])
ttheta_lb = np.array([1.0, 0.35])
ttheta_hb = np.array([1.0, 3.13])

## Labels and figure names
XLABEL = 'Resource Biomass (kg)'
YLABEL = 'Biomass Consumption Rate (kg.yr$^{-1}$)'
FIGNAME = 'figure-parameterisation-willow-consumption-resource.png'

## Evaluate xaxis
xx = np.arange(0, XX[0], 0.01)
xxx = np.stack([xx, np.repeat(XX[1],xx.shape), np.repeat(XX[2],xx.shape)])

## Sample N random parameter sets from the [low,high] ranges
num_samps = 100
tttheta_samps = []    
for i in range(ttheta.shape[0]): 
    ttheta_samps = np.random.uniform(ttheta_lb[i], ttheta_hb[i], num_samps)
    tttheta_samps += [ttheta_samps]
tttheta_samps = np.stack(tttheta_samps)

## Compute the samplecurve values
yy_samps = [y(xxx, tttheta_samps[:,k]) for k in range(num_samps)]

## Compute percentiles
ci95_lower, ci95_upper = np.percentile(yy_samps, [2.5, 97.5], axis=0)
ci50_lower, ci50_upper = np.percentile(yy_samps, [25, 75],   axis=0)
yy_med = np.percentile(yy_samps, 50, axis=0)

## Evaluate
yy_m = y(xxx, ttheta_ct)
    
## Visualise
plt.figure(figsize=FIGSIZE)
# for yy in yy_samps:
#    plt.plot(xx, yy, color='grey', alpha=0.2)
plt.plot(xx, yy_m, label='Midpoint', color='salmon')
plt.plot(xx, yy_med, label='Median', color='grey')
plt.plot(xx, xx, '--', color='black', lw=1.0)
plt.fill_between(xx, ci95_lower, ci95_upper, color='grey', alpha=0.2, label='95% CI')
plt.fill_between(xx, ci50_lower, ci50_upper, color='grey', alpha=0.4, label='50% CI')
plt.xlabel(XLABEL)
plt.ylabel(YLABEL)
# plt.ylim(0,np.max(ci95_upper))
plt.legend(frameon=False, loc='lower right')
plt.savefig(FIGNAME)
plt.show()

## Labels and figure names
XLABEL = 'Population Density (N)'
YLABEL = 'Biomass Consumption Rate (kg.yr$^{-1}$)'
FIGNAME = 'figure-parameterisation-willow-consumption-density.png'

## Evaluate xaxis
xx = np.arange(0, XX[1], 0.01)
xxx = np.stack([np.repeat(XX[0],xx.shape), xx, np.repeat(XX[2],xx.shape)])

## Sample N random parameter sets from the [low,high] ranges
num_samps = 100
tttheta_samps = []    
for i in range(ttheta.shape[0]): 
    ttheta_samps = np.random.uniform(ttheta_lb[i], ttheta_hb[i], num_samps)
    tttheta_samps += [ttheta_samps]
tttheta_samps = np.stack(tttheta_samps)

## Compute the samplecurve values
yy_samps = [y(xxx, tttheta_samps[:,k]) for k in range(num_samps)]

## Compute percentiles
ci95_lower, ci95_upper = np.percentile(yy_samps, [2.5, 97.5], axis=0)
ci50_lower, ci50_upper = np.percentile(yy_samps, [25, 75],   axis=0)
yy_med = np.percentile(yy_samps, 50, axis=0)

## Evaluate
yy_m = y(xxx, ttheta_ct)
    
## Visualise
plt.figure(figsize=FIGSIZE)
# for yy in yy_samps:
#    plt.plot(xx, yy, color='grey', alpha=0.2)
plt.plot(xx, yy_m, label='Midpoint', color='salmon')
plt.plot(xx, yy_med, label='Median', color='grey')
# plt.plot(xx, xx, '--', color='black', lw=1.0)
plt.fill_between(xx, ci95_lower, ci95_upper, color='grey', alpha=0.2, label='95% CI')
plt.fill_between(xx, ci50_lower, ci50_upper, color='grey', alpha=0.4, label='50% CI')
plt.xlabel(XLABEL)
plt.ylabel(YLABEL)
# plt.ylim(0,np.max(ci95_upper))
plt.legend(frameon=False, loc='lower right')
plt.savefig(FIGNAME)
plt.show()

## Labels and figure names
XLABEL = 'Biomass Willow (kg)'
YLABEL = 'Biomass Consumption Rate (kg.yr$^{-1}$)'
FIGNAME = 'figure-parameterisation-willow-consumption-biomass.png'

## Evaluate xaxis
xx = np.arange(0, XX[2], 0.01)
xxx = np.stack([np.repeat(XX[0],xx.shape), np.repeat(XX[1],xx.shape), xx])

## Sample N random parameter sets from the [low,high] ranges
num_samps = 100
tttheta_samps = []    
for i in range(ttheta.shape[0]): 
    ttheta_samps = np.random.uniform(ttheta_lb[i], ttheta_hb[i], num_samps)
    tttheta_samps += [ttheta_samps]
tttheta_samps = np.stack(tttheta_samps)

## Compute the samplecurve values
yy_samps = [y(xxx, tttheta_samps[:,k]) for k in range(num_samps)]

## Compute percentiles
ci95_lower, ci95_upper = np.percentile(yy_samps, [2.5, 97.5], axis=0)
ci50_lower, ci50_upper = np.percentile(yy_samps, [25, 75],   axis=0)
yy_med = np.percentile(yy_samps, 50, axis=0)

## Evaluate
yy_m = y(xxx, ttheta_ct)
    
## Visualise
plt.figure(figsize=FIGSIZE)
# for yy in yy_samps:
#    plt.plot(xx, yy, color='grey', alpha=0.2)
plt.plot(xx, yy_m, label='Midpoint', color='salmon')
plt.plot(xx, yy_med, label='Median', color='grey')
# plt.plot(xx, xx, '--', color='black', lw=1.0)
plt.fill_between(xx, ci95_lower, ci95_upper, color='grey', alpha=0.2, label='95% CI')
plt.fill_between(xx, ci50_lower, ci50_upper, color='grey', alpha=0.4, label='50% CI')
plt.xlabel(XLABEL)
plt.ylabel(YLABEL)
# plt.ylim(0,1)
plt.legend(frameon=False, loc='lower right')
plt.savefig(FIGNAME)
plt.show()

#
###
```

```python
## Notes:
## Find estimate of belowground biomass of willow in kg/m2
## Find estimate of N --> aspen biomass conversion efficiency (currently assuming 100-fold)

## For willow
available_nitrogen = 40 * 1e-6 * 1e4 * 365 # (kg / m^2 / year) Geremia et al. (2025), Fig. 3, C
print(available_nitrogen) # kg / m^2 / year 

## Nitrogen uptake rate of willow (Guo et al. 2025, p. 6(6), Fig. 4)
molar_mass_nitrogen = 14.007
total_daily_nitrogen_accumulation = (184+47+41) * 10**(-6) # (mol N / g willow / day)
uptake_nitrogen_willow = molar_mass_nitrogen * total_daily_nitrogen_accumulation * 365 
print(uptake_nitrogen_willow)
print(uptake_nitrogen_willow/available_nitrogen)
nitrogen_efficiency = (148 + 169 + 148 + 108)/4 # (Weih et al. 2021, p. 4(4), Table 4)
yearly_biomass_consumption_willow = uptake_nitrogen_willow * nitrogen_efficiency
print(yearly_biomass_consumption_willow)

## Nitrogen uptake rate of aspen
weight = 1 * 10**(+3) # weight (g)
shoot_root_ratio = 0.2 # shoot-to-root ratio (prop.)
uptake_per_root_mass = 150 * 10**(-6) # nitrogen uptake per g of root per hour (g N/ g root aspen / hour)
uptake_nitrogen_aspen = weight * shoot_root_ratio * uptake_per_root_mass * 24 * 365 * 10**(-3) # nitrogen uptake per kg per year (kg N / year)
print(uptake_nitrogen_aspen)
print(uptake_nitrogen_aspen/available_nitrogen)
yearly_biomass_consumption_aspen = uptake_nitrogen_aspen * nitrogen_efficiency
print(yearly_biomass_consumption_aspen)

## Nitrogen uptake rate of aspen (alternative calculation)
mass_root_system = 2778 # g/m^2 (Steele et al. 1997, p. 581(5) Table 4.)
uptake_per_root_mass = 150 * 10**(-6) # nitrogen uptake per g of root per hour (g N/ g root aspen / hour)
uptake_nitrogen_aspen = mass_root_system * uptake_per_root_mass * 24 * 365 * 10**(-3) # nitrogen uptake per kg per year (kg N / year / m^2)
print(uptake_nitrogen_aspen)
print(uptake_nitrogen_aspen/available_nitrogen)
yearly_biomass_consumption_aspen = uptake_nitrogen_aspen * nitrogen_efficiency
print(yearly_biomass_consumption_aspen)
```
