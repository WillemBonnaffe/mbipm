```python
#####################
## STANDARDISATION ##
#####################

def compute_nan_mean_std(x, dim=None, keepdim=False, eps=1e-8):
    mask = ~torch.isnan(x)
    count = mask.sum(dim=dim, keepdim=True)

    x_masked = torch.where(mask, x, torch.tensor(0.0, dtype=x.dtype, device=x.device))
    mean = x_masked.sum(dim=dim, keepdim=True) / count

    var = ((torch.where(mask, x - mean, torch.tensor(0.0, dtype=x.dtype, device=x.device))) ** 2).sum(dim=dim, keepdim=True) / count
    std = torch.sqrt(var + eps)

    if not keepdim and dim is not None:
        mean = mean.squeeze(dim)
        std = std.squeeze(dim)

    return mean, std

def apply_standardization(x, mean, std):
    z = (x - mean) / std
    z = torch.where(torch.isnan(x), x, z)  # Preserve original NaNs
    return z

def reverse_standardization(z, mean, std):
    x = z * std + mean
    x = torch.where(torch.isnan(z), z, x)  # Preserve original NaNs
    return x

MEAN, STD = compute_nan_mean_std(time_series_tensor, 0)
time_series_tensor_standardised = apply_standardization(time_series_tensor, MEAN, STD)

## Check
plot_time_series(time_series_tensor_standardised)
# plt.plot(time_series_tensor_standardised)
# plt.yscale('log')
# plt.show()

#
###
```

```python
####################
## DEFINE WRAPPER ##
####################

## Model
model_r = YellowstoneEcosystemModel("parameters-V1-14-2025-09-30.yaml")
model_p = YellowstoneEcosystemModel("parameters-V1-14-2025-09-30.yaml")

## Probabilities to rates
def r2p(x):
    return 1 - 1/(x+1)
def p2r(x):
    return x/(1-x)

## Define wrapper
def wrapper(ttheta, t_max=30, model_r=model_r, model_p=model_p, Dt0=Dt0, Bt0=Bt0, Pt0=Pt0, Gt0=Gt0, NNt0=NNt0, NNNzt0=NNNzt0):
    
    ## Initiate idx
    idx = 0
    
    ## Initial conditions unstructured states
    Dt0_p = Dt0 * ttheta[idx] * 1; idx += 1
    Bt0_p = Bt0 * ttheta[idx] * 1; idx += 1
    Pt0_p = Pt0 * ttheta[idx] * 1; idx += 1
    Gt0_p = Gt0 * ttheta[idx] * 1; idx += 1
    ## Parameter index: 4
    
    ## Initial conditions structured states
    NNt0_p = NNt0.clone()
    NNNzt0_p = NNNzt0.clone()        
    for i in range(len(NNt0_p)):
        NNt0_p[i] = NNt0[i] * ttheta[idx]
        NNNzt0_p[i] = NNNzt0[i] * ttheta[idx]
        idx += 1
    ## Parameter index: 4 + 6

    ## Unstructured variables 
    model_p.phi_BD = r2p(p2r(model_r.phi_BD) * ttheta[idx]); idx += 1 # [0,1] 
    model_p.phi_DB = r2p(p2r(model_r.phi_DB) * ttheta[idx]); idx += 1 # [0,1] 
    model_p.rho_B = model_r.rho_B * ttheta[idx]; idx += 1 # R+ 
    model_p.phi_DP = r2p(p2r(model_r.phi_DP) * ttheta[idx]); idx += 1 # [0,1]         
    model_p.phi_GD = r2p(p2r(model_r.phi_GD) * ttheta[idx]); idx += 1 # [0,1]
    model_p.phi_PG = r2p(p2r(model_r.phi_PG) * ttheta[idx]); idx += 1 # [0,1]
    model_p.rho_G = model_r.rho_G * ttheta[idx]; idx += 1 # R+
    model_p.gamma_3_G = model_r.gamma_3_G * ttheta[idx]; idx += 1 # R+    

    ## Diet preference (I * 4) 
    model_p.lllambda[4][2] = r2p(p2r(model_r.lllambda[4][2]) * ttheta[idx]); idx += 1 # [0,1]
    model_p.lllambda[5][2] = r2p(p2r(model_r.lllambda[5][2]) * ttheta[idx]); idx += 1 # [0,1]
    # Only for scavenging in wolves and cougars
    
    for i in range(len(NNt0_p)):

        ## Nutrient assimilation and grazing (I * 2)
        model_p.bbeta_dh0_P[i] = r2p(p2r(model_r.bbeta_dh0_P[i]) * ttheta[idx]); idx += 1 # [0,1]
        model_p.bbeta_dh0_G[i] = r2p(p2r(model_r.bbeta_dh0_G[i]) * ttheta[idx]); idx += 1 # [0,1]

        ## Predation and scavenging (I * I * 4)
        for j in range(len(NNt0_p)):
            model_p.bbbeta_dp0[i][j] = r2p(p2r(model_r.bbbeta_dp0[i][j]) * ttheta[idx]); idx += 1 # [0,1]
            model_p.bbbeta_dp1[i][j] = model_r.bbbeta_dp1[i][j] * ttheta[idx]; idx += 1 # R
            model_p.bbbeta_dp2[i][j] = model_r.bbbeta_dp2[i][j] * ttheta[idx]; idx += 1 # R       
            # model_p.bbbeta_dn0[i][j] = r2p(p2r(model_r.bbbeta_dn0[i][j]) * ttheta[idx]); idx += 1 # [0,1] 
            # Mask to prevent scavenging where not possible

        ## Natural survival (I * 3)
        model_p.bbeta_sn0[i] = r2p(p2r(model_r.bbeta_sn0[i]) * ttheta[idx]); idx += 1 # [0,1]
        model_p.bbeta_sn1[i] = model_r.bbeta_sn1[i] * ttheta[idx]; idx += 1 # R
        model_p.bbeta_sn2[i] = model_r.bbeta_sn2[i] * ttheta[idx]; idx += 1 # R    

        ## Consumption and growth (I * 4)
        model_p.aalpha_1[i] = r2p(p2r(model_r.aalpha_1[i] * model_r.aalpha_2[i]) * ttheta[idx]); idx += 1 # [0,1]
        model_p.aalpha_2[i] = torch.ones_like(model_p.aalpha_2[i]) # Un-identifiable
        model_p.ggamma_1[i] = model_r.ggamma_1[i] * ttheta[idx]; idx += 1 # R+
        model_p.ggamma_2[i] = r2p(p2r(model_r.ggamma_2[i]) * ttheta[idx]); idx += 1 # [0,1]        

        ## Reproduction (I * 5)
        model_p.bbeta_b0[i] = r2p(p2r(model_r.bbeta_b0[i]) * ttheta[idx]); idx += 1 # [0,1]
        model_p.bbeta_b1[i] = model_r.bbeta_b1[i] * ttheta[idx]; idx += 1
        model_p.bbeta_b2[i] = model_r.bbeta_b2[i] * ttheta[idx]; idx += 1        
        model_p.nnu_o[i] = model_r.nnu_o[i] * ttheta[idx]; idx += 1
        model_p.mmu_o[i] = model_r.mmu_o[i] * ttheta[idx]; idx += 1

        ## Metabolic costs (I * 1)
        model_p.ddelta[i] = model_r.ddelta[i] * ttheta[idx]; idx += 1        

    ## DEBUG (check number of parameters)
    # print(idx)
    
    ## Simulate model
    predictions = model_p.simulate(t_max=t_max, Dt=Dt0_p, Bt=Bt0_p, Pt=Pt0_p, Gt=Gt0_p, NNt=NNt0_p, NNNzt=NNNzt0_p, zzmax=zzmax, zzz_len=zzz_len)

    return predictions

def build_ttheta_index_map(I, J):
    """
    Build a dictionary mapping parameter names to their index in ttheta.
    I = number of species
    J = number of interaction targets per species
    """
    index_map = {}
    idx = 0

    # Initial unstructured state
    for name in ['Dt0', 'Bt0', 'Pt0', 'Gt0']:
        index_map[name] = idx
        idx += 1

    # Initial structured state (NNt0 and NNNzt0)
    for i in range(I):
        index_map[f'NNt0[{i}]'] = idx
        index_map[f'NNNzt0[{i}]'] = idx
        idx += 1

    # Unstructured parameters (9)
    for name in ['phi_BD', 'phi_DB', 'rho_B', 'phi_DP', 'phi_GD', 'phi_PG', 'rho_G', 'gamma_3_G']:
        index_map[name] = idx
        idx += 1        

    ## Diet preference
    index_map[f'lllambda[{4}][{2}]'] = idx; idx +=1
    index_map[f'lllambda[{5}][{2}]'] = idx; idx +=1

    # Nutrient assimilation & grazing (2 per species)
    for i in range(I):
        
        ## Nutrients and grass acquisition
        index_map[f'bbeta_dh0_P[{i}]'] = idx; idx += 1
        index_map[f'bbeta_dh0_G[{i}]'] = idx; idx += 1

        # Predation/scavenging (4 per species pair)
        for j in range(J):
            index_map[f'bbbeta_dp0[{i}][{j}]'] = idx; idx += 1
            index_map[f'bbbeta_dp1[{i}][{j}]'] = idx; idx += 1
            index_map[f'bbbeta_dp2[{i}][{j}]'] = idx; idx += 1
            # index_map[f'bbbeta_dn0[{i}][{j}]'] = idx; idx += 1

        # Natural survival (3 per species)
        index_map[f'bbeta_sn0[{i}]'] = idx; idx += 1
        index_map[f'bbeta_sn1[{i}]'] = idx; idx += 1
        index_map[f'bbeta_sn2[{i}]'] = idx; idx += 1

        # Consumption and growth (4 per species)
        index_map[f'aalpha_1[{i}]'] = idx; idx += 1
        index_map[f'aalpha_2[{i}]'] = 'unidentifiable'  # Not used
        index_map[f'ggamma_1[{i}]'] = idx; idx += 1
        index_map[f'ggamma_2[{i}]'] = idx; idx += 1

        # Reproduction (5 per species)
        index_map[f'bbeta_b0[{i}]'] = idx; idx += 1
        index_map[f'bbeta_b1[{i}]'] = idx; idx += 1
        index_map[f'bbeta_b2[{i}]'] = idx; idx += 1
        index_map[f'nnu_o[{i}]'] = idx; idx += 1
        index_map[f'mmu_o[{i}]'] = idx; idx += 1

        # Metabolic cost (1 per species)
        index_map[f'ddelta[{i}]'] = idx; idx += 1

    return index_map

def get_index_from_name(name, index_map):
    """
    Given a parameter name, return the index in ttheta using the provided map.
    """
    if name not in index_map:
        raise ValueError(f"Parameter name '{name}' not found in index map.")
    return index_map[name]
    
#
###
```

```python
####################
## LAPLACE PRIORS ##
####################

def log_prior(ttheta, sigma):
    return -1.0/sigma * torch.sum(torch.abs(ttheta - 1.0))

ttheta_0 = torch.ones(num_parameters, dtype=torch.float64)
print(log_prior(ttheta_0, .2))
    
#
###
```

```python
################
## LIKELIHOOD ##
################

def log_likelihood(ttheta):

    ## Predict
    predictions = wrapper(ttheta)
    
    ## Format
    time_series_tensor_predicted = format_predicted_time_series(predictions)

    ## Check for nans
    if torch.sum(torch.isnan(time_series_tensor_predicted))>1:
        log_lik = -float('inf')
        # print("check-1") # DEBUG
    else:
        """
        if torch.prod(time_series_tensor_predicted[-1,:]>1.0) == 0.0: # Species extinct

            log_lik = -float('inf')        
            print("check-2") # DEBUG

        else:
        """        
        ## Standardise
        time_series_tensor_predicted_standardised = apply_standardization(format_predicted_time_series(predictions), MEAN, STD)            
        
        ## Compute residuals
        res = time_series_tensor_predicted_standardised - time_series_tensor_standardised

        ## Mean of mean squares for each time series            
        log_lik = torch.tensor(0.0, dtype=torch.float64)
        for i in range(res.shape[1]):
            log_lik += - torch.nanmean(res[:,i]**2)

    ## End
    return log_lik

#
###
```

```python
###############
## POSTERIOR ##
###############

def log_posterior(ttheta, sigma=0.2):
    log_lik = log_likelihood(ttheta)
    log_pri = log_prior(ttheta, sigma)
    log_pos = log_lik + log_pri
    return log_pos, log_lik, log_pri

t0 = time.time()
ttheta_0 = torch.ones(num_parameters, dtype=torch.float64)
print(log_posterior(ttheta_0))
tf = time.time()
print(f"time: {tf-t0:2f}s")
    
#
###
```

```python
############
## TARGET ##
############

SIGMA = 1.0
def d_target(ttheta):    
    # ttheta = torch.exp(ttheta) # Search only on strictly positive real space
    log_pos, _, _ = log_posterior(ttheta, sigma=SIGMA)
    return log_pos

#
###
```

```python
#########################################
## FUNCTIONS ANALYSIS OF CHAIN RESULTS ##
#########################################

def get_map(chain):
    idx = torch.argmax(chain[:,0])
    ttheta_map = chain[idx,1:]
    return ttheta_map

def plot_chain(chain):
    """
    """
    # Trace plots
    plt.figure()
    plt.plot(chain[:, 0])
    # plt.title("Trace plot: theta[0]")
    plt.xlabel("Iteration")
    plt.ylabel("Value")
    plt.show()
    
    plt.figure()
    plt.plot(chain[:, 1])
    # plt.title("Trace plot: theta[0]")
    plt.xlabel("Iteration")
    plt.ylabel("Value")
    plt.show()
    
    plt.figure()
    plt.plot(chain[:, 2])
    # plt.title("Trace plot: theta[1]")
    plt.xlabel("Iteration")
    plt.ylabel("Value")
    plt.show()
    
    # Scatter plot
    plt.figure()
    plt.plot(chain[:, 1], chain[:, 2])
    # plt.scatter(samples[:, 0], samples[:, 1], alpha=0.3, s=1)
    # plt.title("Posterior samples")
    plt.xlabel("theta[0]")
    plt.ylabel("theta[1]")
    # plt.axis("equal")
    plt.show()
    
    plt.figure()
    plt.plot(chain[:,1:11], alpha=0.2)
    plt.show()

def plot_fit(ttheta_0, ttheta_map):
    """
    """
    ## Predict
    predictions_0 = wrapper(ttheta_0)
    time_series_tensor_predicted_0 = format_predicted_time_series(predictions_0)
    
    ## Predict
    predictions_map = wrapper(ttheta_map)
    time_series_tensor_predicted = format_predicted_time_series(predictions_map)
    
    ## Standardise
    time_series_tensor_predicted_standardised_0 = apply_standardization(format_predicted_time_series(predictions_0), MEAN, STD)
    time_series_tensor_predicted_standardised = apply_standardization(format_predicted_time_series(predictions_map), MEAN, STD)
    time_series_tensor_standardised = apply_standardization(time_series_tensor, MEAN, STD)
    
    ## Visualise
    for i in range(I+1):
        plt.figure(figsize=(3,3))
        plt.plot(time_series_tensor_standardised[:,i],'o-')
        plt.plot(time_series_tensor_predicted_standardised_0[:,i], label='Before training')
        plt.plot(time_series_tensor_predicted_standardised[:,i], label='After training')
        plt.xlabel('Time Step (years)')
        plt.ylabel('Value (standardised)')
        plt.legend()
        plt.show()

def plot_fit_compact(ttheta_0, ttheta_map, I=6):
    """
    Plot predictions before and after training using subplots for compact layout.
    """
    # Predict
    predictions_0 = wrapper(ttheta_0)
    predictions_map = wrapper(ttheta_map)
    
    # Format and standardize
    pred_0 = apply_standardization(format_predicted_time_series(predictions_0), MEAN, STD)
    pred_map = apply_standardization(format_predicted_time_series(predictions_map), MEAN, STD)
    true = apply_standardization(time_series_tensor, MEAN, STD)

    # Layout: Calculate rows and columns for subplots
    cols = 3
    rows = (I + 1 + cols - 1) // cols  # Ceiling division

    fig, axes = plt.subplots(rows, cols, figsize=(cols * 4, rows * 3), constrained_layout=True)
    axes = axes.flatten()

    for i in range(I + 1):
        ax = axes[i]
        ax.plot(true[:, i], 'o-', label='True')
        ax.plot(pred_0[:, i], label='Before training')
        ax.plot(pred_map[:, i], label='After training')
        ax.set_xlabel('Time Step (years)')
        ax.set_ylabel('Value (standardised)')
        ax.set_title(f'Series {i}')
        ax.legend()

    # Turn off any unused subplots
    for j in range(I + 1, len(axes)):
        axes[j].axis('off')

    plt.show()

def plot_model_alt(predictions):

    ## Format output
    zzz, tt, DDt, BBt, PPt, GGt, NNNt, NNNNzt, ttotal_biomass = predictions
    NNNNzt_new = [] # (time, species, phenotype) --> (species, time, phenotype)
    for i in range(6):
        NNNzt_new = []
        for NNNzt in NNNNzt:
            NNNzt_new += [NNNzt[i].clone()]
        NNNNzt_new += [torch.stack(NNNzt_new)]
    NNNNzt = torch.stack(NNNNzt_new)

    ## FIGURE 2    
    for i in range(6):

        ## Visualise matrices
        plt.figure(figsize=(9, 3))
        #
        plt.plot(NNNNzt[i,:].sum(1), color='salmon', label='Predicted')
        if i > 1:
            plt.plot(time_series_tensor[:, i+1], label='Observed')
        # plt.colorbar()
        # plt.title('$N^{(i)}(t)$')
        plt.xlabel('Time (years)')
        plt.ylabel('Number of individuals')
        plt.legend(frameon=False)
        #
        plt.tight_layout()
        plt.show()
        
#
###
```

```python
####################
## INITIATE CHAIN ##
####################

## Import module
# from f_demc import de_mcmc_single_chain
from f_demc_sparse_noise import de_mcmc_single_chain_sparse
from f_demc_sparse_noise import de_mcmc_single_chain_sparse_optimiser

#
###
```

```python
###############
## RUN CHAIN ##
###############

DATE = '2025-10-06'
RUN_ID = f"C1-real-map-sigma-{SIGMA}-{DATE}"
#
## Multiple runs
num_runs = 10
for k in range(2, num_runs):
#
## Selected runs
# for k in [1]:

    print('')
    print(f"{RUN_ID}-{k+1}")    
    
    ## Select initial value
    d_target_ = -torch.inf
    while d_target_ == -torch.inf:
        ttheta_map = ttheta_0 = torch.ones(num_parameters, dtype=torch.float64) + torch.rand(ttheta.shape) * 1e-3
        d_target_ = d_target(ttheta_0)
        print(d_target_)    
    
    ## Initiate run
    num_epoch = 1000
    stop_counter = 0
    max_stop_count = 3
    lloss = [log_posterior(ttheta_map, sigma=SIGMA)]
    cchain = []
    for epoch in range(num_epoch):

        ##
        print('')
        print(f"epoch {epoch+1}/{num_epoch}")
        
        ## Step
        t0 = time.time()
        results = de_mcmc_single_chain_sparse_optimiser(
            d_target=d_target,
            theta_0=ttheta_map,
            epsilon=0.0,
            p_sparse=1e-01,
            scale_sparse=1e-02,
            memory=25,
            n_it=100,
            verbose=True  # set True for intermediate output
        )
        tf = time.time()
        print(f"time: {tf-t0:2f}s")
        
        ## Extract chain
        chain = results['chain']
        cchain += [chain[-1]]
        
        ## Get MaP
        ttheta_map_new = get_map(chain)
        loss = log_posterior(ttheta_map_new, sigma=SIGMA)
        lloss += [log_posterior(ttheta_map_new, sigma=SIGMA)]
        print(loss)
        
        ## Update map
        if d_target(ttheta_map_new) > d_target(ttheta_map) + 1e-4:
            ttheta_map = ttheta_map_new.clone()
            stop_counter = 0 # reset stop counter
        else:
            stop_counter += 1
            if stop_counter >= max_stop_count:            
                print(f"training ended after {epoch+1} epochs")
                break        
                
    ## Save
    torch.save(chain, f"{RUN_ID}-{k+1}-chain.pt")
    torch.save(cchain, f"{RUN_ID}-{k+1}-cchain.pt")
    torch.save(lloss, f"{RUN_ID}-{k+1}-lloss.pt")    
                
#
###
```

```python
###########################
## VISUALISE UNCERTAINTY ##
###########################

# --- Parameters ---
burnin = 0
size_after_thin = 10

# --- Burn-in ---
# chain_burned = chain[1000:4000]
chain_burned = chain[burnin:]

# --- Thinning (uniform) ---
n_remaining = chain_burned.shape[0]
indices = torch.linspace(0, n_remaining - 1, steps=size_after_thin).long()
chain_thinned = chain_burned[indices]
samples_thinned = chain_thinned[:, 1:]

# --- Evaluate and store predictions ---
predictions_list = []
for theta in samples_thinned:
    # theta = torch.exp(theta) # Link function
    pred = wrapper(theta)
    formatted = apply_standardization(format_predicted_time_series(pred), MEAN, STD)
    predictions_list.append(formatted)

predictions_tensor = torch.stack(predictions_list)  # shape: (size_after_thin, T, D)

# --- Compute statistics ---
mean_prediction = predictions_tensor.mean(dim=0)
median_prediction = predictions_tensor.median(dim=0).values
ci50_lower = torch.quantile(predictions_tensor, 0.25, dim=0)
ci50_upper = torch.quantile(predictions_tensor, 0.75, dim=0)
ci95_lower = torch.quantile(predictions_tensor, 0.025, dim=0)
ci95_upper = torch.quantile(predictions_tensor, 0.975, dim=0)

#
###
```

```python
###########################
## VISUALISE UNCERTAINTY ##
###########################

import matplotlib.pyplot as plt

T, D = mean_prediction.shape

for i in range(D):
    plt.figure(figsize=(6, 4))

    # Plot observed data
    plt.plot(time_series_tensor_standardised[:, i], 'o-', label='Observed', color='black', alpha=0.7)

    # Plot mean prediction
    plt.plot(mean_prediction[:, i], label='Mean prediction', color='salmon')

    # Plot median prediction
    plt.plot(median_prediction[:, i], label='Median prediction', color='salmon', linestyle='--')

    # Fill 95% CI
    plt.fill_between(
        torch.arange(T),
        ci95_lower[:, i],
        ci95_upper[:, i],
        color='lightgray',
        alpha=0.2,
        label='95% CI'
    )

    # Fill 50% CI
    plt.fill_between(
        torch.arange(T),
        ci50_lower[:, i],
        ci50_upper[:, i],
        color='gray',
        alpha=0.3,
        label='50% CI'
    )

    plt.xlabel('Time Step')
    plt.ylabel('Standardised Value')
    # plt.title(f'Time Series Dimension {i}')
    plt.legend()
    plt.tight_layout()
    plt.show()

#
###
```
