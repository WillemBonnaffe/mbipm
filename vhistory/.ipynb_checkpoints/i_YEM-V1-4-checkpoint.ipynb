# Yellowstone Ecosystem Model

## Overview
This Jupyter Notebook presents the Yellowstone Ecosystem Model (YEM).
The model contains a simple nutrient recycling model, as well as five species structured by body mass.
The five species are aspen, elk, bison, wolves, and mountain lions.

```python
##############
## INITIATE ##
##############

## Imports
import torch
import numpy as np
import matplotlib.pyplot as plt
import time
import seaborn as sns

## Set the device to MPS if available
device = torch.device("mps" if torch.has_mps else "cpu")
device = "cpu" # Force cpu for now
print(device)

## Figure parameters
plt.rcParams['figure.dpi'] = 100
figsize = (14,5)
custom_params = {"axes.spines.right": False, "axes.spines.top": False, 'axes.facecolor':'whitesmoke'}
sns.set_theme(style="ticks", rc=custom_params)

#
###
```

```python
#######################
## UTILITY FUNCTIONS ##
#######################

def hh(x, y):
  """
  Goal: Compact version of the half-harmonic function.
  """
  return (x * y)/(x + y)

def f_normal(z, mu, sigma):
  """
  Goal: Normal probability density distribution
  Inputs:
  - z (tensor): Variable.
  - mu (float): Mean parameter.
  - sigma (float): Standard deviation parameter.
  Outputs:
  - density (tensor): Probability density.
  """
  prefactor = 1 / (sigma * torch.sqrt(torch.tensor(2 * torch.pi)))
  exponent = -0.5 * ((z - mu) / sigma) ** 2
  density = prefactor * torch.exp(exponent)
  return density

def pdf(x):
  """
  """
  return x/x.sum(0)

#
###
```

## Ecosystem dynamics

We model the dynamics of the ecosystem as follows:

\begin{aligned}
\Delta D(t) &= \phi_{BD} B(t) + \phi_{GD} G(t) - hh(\phi_{DB} D(t), \rho_B B(t)) \phantom{\int} \\
\Delta B(t) &= (1 - \phi_{DP}) \cdot hh(\phi_{DB} D(t), \rho_B B(t)) - \phi_{BD} B(t)\phantom{\int} \\
\Delta P(t) &= \phi_{DP} \cdot hh(\phi_{DB} D(t), \rho_B B(t)) - hh(\phi_{PG} P(t), \rho_G G(t)) \phantom{\int} \\
\Delta G(t) &= hh(\phi_{PG} P(t), \rho_G G(t)) - \phi_{GD} G(t). \phantom{\int} \\
\end{aligned}

where $D$ is the biomass of dead organic matter, $B$ is the biomass of bacteria communities and other decomposers, $P$ is the biomass of nutrients (e.g. phosphates or nitrogen), $G$ is the biomass of grass.
The parameters $\phi \in [0, 1]$ denote fractions, while the parameters $\rho \in \mathbb{R}^{+}$ denote rates.
The function $hh(x,y) = x y / (x + y)$ is the half harmonic function, which is equivalent to the Michaelis-Menten equation.
This function ensures that the biomass consumed cannot exceed the biomass available for consumption, and reduces the per-capita consumption with the density of consumers.
The parameters $\phi_{BD}$ and $\phi_{GD}$ correspond to the proportion of decomposer and grass biomass that is converted into dead organic matter.
The parameters $\phi_{DB}$ and $\rho_{B}$ correspond to the proportion of the biomass of decomposed organic matter available to decomposers and the rate at which it is consumed, respectively.
Similarly, $\phi_{PG}$ and $\rho_{G}$ correspond to the nutrient biomass available for consumption by grass and its consumption rate.



```python
#########################
## ECOSYSTEM FUNCTIONS ##
#########################

def DBPG_tdt(D_t, B_t, P_t, G_t, phi_BD, phi_GD, phi_DP, phi_DB, rho_B, phi_PG, rho_G, P_c, G_c, Z_e_t, Z_d_t):
  """
  Goal: Compute dynamics of the ecosystem.
  Inputs:
  - D_t (tensor): Dead organic matter biomass (kg).
  - B_t (tensor): Decomposer biomass (kg).
  - P_t (tensor): Nutrient biomass (kg).
  - G_t (tensor): Grass biomass (kg).
  - phi_BD (float): Proportion of decomposer biomass converted into dead organic matter (proportion).
  - phi_GD (float): Proportion of grass biomass converted into dead organic matter (proportion).
  - phi_DP (float): Proportion of dead organic matter biomass converted into nutrients by decomposers (proportion).
  - phi_DB (float): Proportion of dead organic matter available to decomposers (proportion).
  - rho_B (float): rate of consumption of decomposers (kg kg^{-1} year^{-1}).
  - phi_PG (float): Proportion of nutrients available to grass (proportion).
  - rho_G (float): rate of consumption of grass (kg kg^{-1} year^{-1}).
  - P_c (float): Nutrient biomass consumed by browse (kg).
  - G_c (float): Grass biomass consumed by herbivores (kg).
  - Z_e_t (float): Biomass excreted by organisms (kg).
  - Z_d_t (float): Biomass of dead organisms (kg).
  Outputs:
  - D_tdt (tensor): Change in dead organic matter biomass (kg).
  - B_tdt (tensor): Change in decomposer biomass (kg).
  - P_tdt (tensor): Change in nutrient biomass (kg).
  - G_tdt (tensor): Change in grass biomass (kg).
  """
  D_tdt = D_t + phi_BD * B_t + phi_GD * G_t - hh(phi_DB * D_t, rho_B * B_t) + Z_d_t + Z_e_t
  B_tdt = B_t + (1 - phi_DP) * hh(phi_DB * D_t, rho_B * B_t) - phi_BD * B_t
  P_tdt = P_t + phi_DP * hh(phi_DB * D_t, rho_B * B_t) - hh(phi_PG * P_t, rho_G * G_t) - P_c
  G_tdt = G_t + hh(phi_PG * P_t, rho_G * G_t) - phi_GD * G_t - G_c
  return D_tdt, B_tdt, P_tdt, G_tdt

#
###
```

```python
################################
## ECOSYSTEM MODEL PARAMETERS ##
################################

s = 1/1000 # Scale of the system (10^3 kg)

## Ecosystem parameters (Set 1)
phi_BD = 0.05 # Yearly proportion of dead decomposers (prop.)
phi_GD = 0.05 # Yearly proportion of dead grass (prop.)
phi_DP = 0.4 # Yearly proportion of organic matter recycled (prop.)
phi_DB = 0.8 # Yearly proportion of DOM available to bacteria (prop.)
rho_B = 365 * 0.0004 # Yearly DOM consumption of 1 kg of bacteria (kg kg^{-1} year^{-1})
phi_PG = 0.8 # Yearly proportion of nutrients available to grass (prop.)
rho_G = 365 * 0.0125 # Yearly nutrient consumption of 1 kg of grass (kg kg^{-1} year^{-1})

#
###

################
## SIMULATION ##
################

## Parameters of simulation
t_max = 101
times = range(0, t_max)

## Initiate vectors
D = torch.zeros(t_max)
B = torch.zeros(t_max)
P = torch.zeros(t_max)
G = torch.zeros(t_max)

## Initiate states
D[0] = D_t = 1*10**6
B[0] = B_t = 1*10**6
P[0] = P_t = 1*10**6
G[0] = G_t = 1*10**6

## Run model
for time in range(0,t_max-1):

  ## Next state
  D_t, B_t, P_t, G_t = DBPG_tdt(D_t, B_t, P_t, G_t, phi_BD, phi_GD, phi_DP, phi_DB, rho_B, phi_PG, rho_G, P_c=0, G_c=0, Z_e_t=0, Z_d_t=0)

  ## Collect
  D[time+1] = D_t
  B[time+1] = B_t
  P[time+1] = P_t
  G[time+1] = G_t

## Compute total biomass
total_biomass = D + B + P + G

#
###

###################
## VISUALISATION ##
###################

## Figure
plt.figure(figsize=(12,4))

## Define color vector
color_vector = {
  'Total_biomass': 'black',
  'Dead organic matter': 'brown',
  'Decomposers': 'blue',
  'Nutrients': 'orange',
  'Grass': 'green'
}

## Ecosystem dynamics
plt.subplot(1,2,1)
plt.plot(times, total_biomass, label='Total_biomass', color=color_vector['Total_biomass'])
plt.plot(times, D, label='Dead organic matter', color=color_vector['Dead organic matter'])
plt.plot(times, B, label='Decomposers', color=color_vector['Decomposers'])
plt.plot(times, P, label='Nutrients', color=color_vector['Nutrients'])
plt.plot(times, G, label='Grass', color=color_vector['Grass'])
plt.legend(loc='upper right')
plt.xlabel('Time')
plt.ylabel('Biomass (kg)')
plt.grid(c='white')

## Calculate proportions
total_biomass = D + B + P + G
D_prop = D / total_biomass
B_prop = B / total_biomass
P_prop = P / total_biomass
G_prop = G / total_biomass

## Convert tensors to numpy for plotting
proportions = torch.stack([D_prop, B_prop, P_prop, G_prop], dim=0).numpy()

## Stacked area plot
plt.subplot(1,2,2)
labels = ['Dead organic matter', 'Decomposers', 'Nutrients', 'Grass']
colors = [color_vector[label] for label in labels]  # Extract colors for each label
plt.stackplot(times, proportions, labels=labels, colors=colors, alpha=0.7)

## Remove gap between plot and y-axis
plt.xlim(0, t_max-1)
plt.ylim(0,1)

## Customize plot
plt.legend(loc='upper right')
plt.xlabel('Time')
plt.ylabel('Proportion of Total Biomass')
# plt.title('Change in Biomass Proportions Over Time')
plt.grid(c='white')
plt.show()

#
###
```


## Population dynamics of the i-th species

In addition, we introduce a model for $I$ populations structured by body mass to incorporate aspects of bioenergetics and ontogeny key to the dynamics of real systems.
The model describing the general population dynamics is of the form:

\begin{equation}
N^{(i)}(z, t + 1) = \int_{S_z} \left \{ f_r^{(i)}(z | z') n_{o}^{(i)}(z') p_{b}^{(i)}(z') + f_s^{(i)}(z | z') \right \} p_s^{(i)}(z') N^{(i)}(z', t) dz',
\end{equation}

where $N^{(i)}(z, t)$ is density of individuals with phenotype $z$ at time $t$.
$S_z$ is the support of the phenotype.
$f_r^{(i)}(z|z')$ and $f_s^{(i)}(z|z')$ are the probability density function or recruitment and survival to $z$ given $z'$, respectively.
The term $n_{o}^{(i)}(z')$ is the average number of offspring produced by a single individual with phenotype $z'$.
The terms $p_{b}^{(i)}(z')$, and $p_s^{(i)}(z')$ correspond to the probability (or proportion) of breeding, and surviving from time $t$ to $t+1$.



## Phenotype

We introduce the relationship between the phenotype and the population density.
The phenotype is the body mass of individuals, $z$, in kg.
The total biomass in the population can be obtained by summing the body size of the all individuals in the population,

\begin{equation}
Z^{(i)}(t) = \int z N^{(i)}(z,t) dz.
\end{equation}

In our current framework, we impose the following two mechanistic constraints:
(1) Biomass conservation through reproduction: individuals use their own reserves to produce offspring, which means that though total population density can increase, no increase in population biomass is possible through reproduction.
(2) Density conservation through growth: conversely, gains in body mass through ontogeny should increase total population biomass but should not affect total population density, as inviduals gain weight but do not multiply.

## Overall survival

Predators have a direct impact on the populations by contributing to the mortality of individuals.
We separate survival into an intrinsic component and a predation component.
We can do this by assuming that individuals need to survive both natural causes of mortality and predation:

\begin{aligned}
p_s^{(i)}(z) &= p_{s,n}^{(i)}(z) p_{s,p}^{(i)}(z) \\
\Leftrightarrow p_d^{(i)}(z) &= 1 - p_{s,n}^{(i)}(z) p_{s,p}^{(i)}(z), \\
\end{aligned}

where $p_{s,n}^{(i)}$ and $p_{s,p}^{(i)}$ denote the probability that an individual from species $i$ survives natural causes of death and predation, respectively, to the next year.


## Intrinsic survival

We model individual intrinsic survival as a sigmoid function of body mass,

\begin{equation}
p_{s,n}^{(i)}(z) = \beta_{s,n,0}^{(i)} \cdot \mathrm{sigmoid} \left\{\left(z - \beta_{s,n,1}^{(i)}\right) \cdot \beta_{s,n,2}^{(i)} \right\},
\end{equation}

where $\beta_{s,n,0}^{(i)}$ is maximum survival probability, and $\beta_{s,n,1}^{(i)}$ is the body size at maximum increase in log-odds of surviving per unit of phenotype, $\beta_{s,n,2}^{(i)}$ is the slope.
Breeding is defined in the same way with a specific set of parameters


```python
####################################
## INTRINSIC SURVIVAL PROBABILITY ##
####################################

## Define function using PyTorch
def intrinsic_survival_i_j(z, beta_0, beta_1, beta_2):
    return beta_0 * torch.sigmoid((z - beta_1) * beta_2)

## Define parameter sets for visualization
param_sets = [
    {"beta_0": 0.98, "beta_1": -1.0, "beta_2": 1.0, "label": "Aspen"},
    {"beta_0": 0.96, "beta_1": 0.0, "beta_2": 0.025, "label": "Elk"},
    {"beta_0": 0.9, "beta_1": 0.0, "beta_2": 0.025, "label": "Bison"},
    {"beta_0": 1-1/12, "beta_1": 0.0, "beta_2": 0.025, "label": "Wolves"},
    {"beta_0": 1-1/12, "beta_1": 0.0, "beta_2": 0.04, "label": "Cougars"},
]
## Range of body mass values (converted to PyTorch tensor)
z_values = torch.linspace(1, 1000, 1000)

## Plot survival functions for different parameter sets
plt.figure(figsize=(6, 4))
for params in param_sets:
    y = intrinsic_survival_i_j(z_values, params["beta_0"], params["beta_1"], params["beta_2"])
    plt.plot(z_values.numpy(), y.numpy(), label=params["label"])

## Formatting the plot
plt.xlabel("Body Mass (z) (kg)")
plt.ylabel("Intrinsic Survival Probability")
# plt.title("Intrinsic Survival as a Function of Body Mass")
plt.legend(fontsize=8, loc="upper left", bbox_to_anchor=(1, 1))
plt.xscale("log")
# plt.grid(c='white')

## Show the plot
plt.show()

#
###
```

## Survival to predation

Similarly, we assume that surviving predation by $J$ different species is given by

\begin{aligned}
p_{s,p}^{(i)}(z) &= \prod_{j=1}^{J} p_{s,p}^{(i,j)}(z). \\
\end{aligned}

We model the probability of dying from predation as a geometric series (i.e. decreasing probability of survival with the number of encounters with individual predators):

\begin{equation}
p_{s,p}^{(i,j)}(z) = \prod_{k=1}^{N^{(j)}(t)}p_{s,p}^{(i,j,k)}(z),
\end{equation}

where $p_{s,p}^{(i,j,k)}(z)$ is the probability that an individual of species $i$ survives an encounter with a single predator $k$ of species $j$, and $N^{(j)}(t)$ is the number of predators of species $j$ present at time $t$.
This is equivalent to considering the point of view of a single predator and assuming that it is capable of killing the prey with probability $p_{d,p}^{(i,j,k)}$ (see proof in supplementary).
In practice, we compute the probability of surviving predation as

\begin{equation}
p_{s,p}^{(i,j)}(z) = \exp \left\{ N^{(j)}(t) \cdot \log \left\{ p_{s,p}^{(i,j,k)}(z) \right\}\right\}, \phantom{\int}
\end{equation}

to increase numerical stability of multiplying many values.
Given that mountain lions and wolves prey on younger individuals, we introduce a decrease in the probability of predation with body mass, using a sigmoid function:

\begin{aligned}
p_{s,p}^{(i,j,k)}(z) &= 1 - \beta_{d,p,0}^{(i,j)} \cdot \mathrm{sigmoid} \left\{\left(z - \beta_{d,p,1}^{(i,j)}\right) \cdot \beta_{d,p,2}^{(i,j)}\right\}, \\
\end{aligned}

where the parameters $\beta_{d,p,0}^{(i,j)}$ denote the maximum probability of death of individual of species $i$ by predation by individual of species $j$, $\beta_{d,p,1}^{(i,j)}$ is the body mass at which prey escape predation, and $\beta_{d,p,2}^{(i,j)}$ is the slope, which controls how sharp the transition from predated to not predated is.


```python
####################################
## EXTRINSIC SURVIVAL PROBABILITY ##
####################################

## Define function using PyTorch
def extrinsic_survival_i_j(z, beta_0, beta_1, beta_2, N_j):
    return torch.exp(N_j * torch.log((1 - beta_0 * torch.sigmoid((z - beta_1) * beta_2))))

## Define parameter sets for visualization
param_sets = [
    {"N_j": 100, "beta_0": 0.01, "beta_1": 2.4, "beta_2": -1.0, "label": "Aspen by Elk"},
    {"N_j": 100, "beta_0": 0.001, "beta_1": 150.0, "beta_2": -0.025, "label": "Elk by Wolf"},
    {"N_j": 100, "beta_0": 0.002, "beta_1": 150.0, "beta_2": -0.025, "label": "Elk by Cougar"},
    {"N_j": 100, "beta_0": 0.0, "beta_1": -1.0, "beta_2": 1.0, "label": "Bison by Wolf"},
    {"N_j": 100, "beta_0": 0.0, "beta_1": -1.0, "beta_2": 1.0, "label": "Bison by Cougar"},
    {"N_j": 100, "beta_0": 0.001, "beta_1": 50.0, "beta_2": -0.025, "label": "Wolves by Wolf"},
    {"N_j": 100, "beta_0": 0.001, "beta_1": 50.0, "beta_2": -0.025, "label": "Wolves by Cougar"},
    {"N_j": 100, "beta_0": 0.001, "beta_1": 50.0, "beta_2": -0.025, "label": "Cougars by Wolf"},
    {"N_j": 100, "beta_0": 0.001, "beta_1": 50.0, "beta_2": -0.025, "label": "Cougars by Cougar"},
]
## Range of body mass values (converted to PyTorch tensor)
z_values = torch.linspace(1, 1000, 1000)

## Plot survival functions for different parameter sets
plt.figure(figsize=(6, 4))
for params in param_sets:
    y = extrinsic_survival_i_j(z_values, params["beta_0"], params["beta_1"], params["beta_2"], params["N_j"])
    plt.plot(z_values.numpy(), y.numpy(), label=params["label"])

## Formatting the plot
plt.xlabel("Body Mass (z) (kg)")
plt.ylabel("Extrinsic Survival Probability")
plt.legend(fontsize=8, loc="upper left", bbox_to_anchor=(1, 1))
plt.xscale("log")
# plt.grid(c='white')

## Show the plot
plt.show()

#
###
```

```python
########################
## SURVIVAL FUNCTIONS ##
########################

def p_s_i(z, beta_sn0_i, beta_sn1_i, beta_sn2_i, NNt, bbeta_dp0_i, bbeta_dp1_i, bbeta_dp2_i):
    p_s_i_ = p_sn_i(z, beta_sn0_i, beta_sn1_i, beta_sn2_i) * p_sp_i(z, NNt, bbeta_dp0_i, bbeta_dp1_i, bbeta_dp2_i)
    return p_s_i_

def p_sn_i(z, beta_sn0_i, beta_sn1_i, beta_sn2_i):
    p_sn_i_ = beta_sn0_i * torch.sigmoid((z - beta_sn1_i) * beta_sn2_i)
    return p_sn_i_

def p_sp_i(z, NNt, bbeta_dp0_i, bbeta_dp1_i, bbeta_dp2_i):
    p_sp_i_ = 1.0
    for j in range(len(NNt)):
      p_sp_i_ = p_sp_i_ * p_sp_ij(z, NNt[j], bbeta_dp0_i[j], bbeta_dp1_i[j], bbeta_dp2_i[j])
    return p_sp_i_

def p_sp_ij(z, Nt_j, beta_dp0_ij, beta_dp1_ij, beta_dp2_ij):
    p_sp_ij_ = torch.exp(Nt_j * torch.log(p_sp_ijk(z, beta_dp0_ij, beta_dp1_ij, beta_dp2_ij)))
    return p_sp_ij_

def p_sp_ijk(z, beta_dp0_ij, beta_dp1_ij, beta_dp2_ij):
    p_sp_ijk_ = 1 - beta_dp0_ij * torch.sigmoid((z - beta_dp1_ij) * beta_dp2_ij)
    return p_sp_ijk_

def total_p_dp_i(z, NNt, bbeta_dp0_i, bbeta_dp1_i, bbeta_dp2_i):
    p_dp_i_ = 0.0
    for j in range(len(NNt)):
      p_dp_i_ = p_dp_i_ + (1 - p_sp_ij(z, NNt[j], bbeta_dp0_i[j], bbeta_dp1_i[j], bbeta_dp2_i[j]))
    return p_dp_i_

#
###
```

## Partitioning mortality of structured species

In order to know what predators feed on we need to partition the causes of death of individuals into the different species.
The probability of death be expanded to obtain additive contributions of predators, as follows

\begin{aligned}
p_{s}^{(i)}(z) &= p_{s,n}^{(i)}(z) p_{s,p}^{(i)}(z) \\
\Leftrightarrow p_{d}^{(i)}(z) &= 1 - p_{s,n}^{(i)}(z) p_{s,p}^{(i)}(z) \\
&= p_{d,n}^{(i)} + p_{s,n}^{(i)}(z) p_{d,p}^{(i)}(z) \\
&= p_{d,n}^{(i)} + p_{s,n}^{(i)}(z) p_{d,p}^{(i)}(z) \left( \frac{\sum_{j=1}^{I} p_{d,p}^{(i,j)}(z)}{\sum_{k=1}^{I} p_{d,p}^{(i,k)}(z)} \right) \\
&= p_{d,n}^{(i)} + p_{s,n}^{(i)}(z) p_{d,p}^{(i)}(z) \sum_{j=1}^{I} \frac{p_{d,p}^{(i,j)}(z)}{\sum_{k=1}^{I} p_{d,p}^{(i,k)}(z)} \\
\end{aligned}

We can separate the biomass of dead individuals into a natural- and a predation-induced component:

\begin{aligned}
Z_{d}^{(i)}(t) &= \int z' p_d^{(i)}(z') N^{(i)}(z', t) dz' \\
&= \int z' \left( p_{d,n}^{(i)}(z') + p_{s,n}^{(i)}(z') p_{d,p}^{(i)}(z') \right) N^{(i)}(z', t) dz' \\
&= \int z' p_{d,n}^{(i)}(z') N^{(i)}(z', t) dz' + \int z' p_{s,n}^{(i)}(z') p_{d,p}^{(i)}(z') N^{(i)}(z', t) dz' \\
&= Z_{d,n}^{(i)}(t) + Z_{d,p}^{(i)}(t). \phantom{\int} \\
\end{aligned}

This allows us to partition the dead biomass into additive components attributable to each predators,

\begin{aligned}
Z_{d}^{(i)}(t) &= Z_{d,n}^{(i)}(t) + \sum_{j=1}^{J} Z_{d,p}^{(i,j)}(t), \phantom{\int} \\
\end{aligned}

where the term $Z_{d,p}^{(i,j)}(t)$ corresponds to the biomass of species $i$ killed by predation by species $j$ and can be computed as

\begin{aligned}
Z_{d,p}^{(i,j)}(t) &= \int z' p_{s,n}^{(i)}(z') p_{d,p}^{(i)}(z') \frac{p_{d,p}^{(i,j)}(z')}{\sum_{k=1}^{I} p_{d,p}^{(i,k)}(z')} N^{(i)}(z', t) dz'. \phantom{\int} \\
\end{aligned}


```python
########################
## PARTITIONING DEATH ##
########################

def Zt_d_i(zz, NNt_i, beta_sn0_i, beta_sn1_i, beta_sn2_i, NNt, bbeta_dp0_i, bbeta_dp1_i, bbeta_dp2_i):
    Zt_d_i_ = zz * (1 - p_s_i(zz, beta_sn0_i, beta_sn1_i, beta_sn2_i, NNt, bbeta_dp0_i, bbeta_dp1_i, bbeta_dp2_i)) * NNt_i
    return Zt_d_i_.sum()

def Zt_dn_i(zz, NNt_i, beta_sn0_i, beta_sn1_i, beta_sn2_i):
    Zt_dn_i_ = zz * (1 - p_sn_i(zz, beta_sn0_i, beta_sn1_i, beta_sn2_i)) * NNt_i
    return Zt_dn_i_.sum()

def ZZt_dp_i(zz, NNt_i, beta_sn0_i, beta_sn1_i, beta_sn2_i, NNt, bbeta_dp0_i, bbeta_dp1_i, bbeta_dp2_i):
    p_sp_i_ = p_sp_i(zz, NNt, bbeta_dp0_i, bbeta_dp1_i, bbeta_dp2_i)
    total_p_dp_i_ = total_p_dp_i(zz, NNt, bbeta_dp0_i, bbeta_dp1_i, bbeta_dp2_i)
    ZZt_dp_i_ = []
    for j in range(len(NNt)):
      if total_p_dp_i_.sum() == 0:
        ZZt_dp_i_.append(0)
      else:
        ZZt_dp_i_.append(
            (zz * p_sn_i(zz, beta_sn0_i, beta_sn1_i, beta_sn2_i) * (1 - p_sp_i_) * (1 - p_sp_ij(zz, NNt[j], bbeta_dp0_i[j], bbeta_dp1_i[j], bbeta_dp2_i[j])) / total_p_dp_i_ * NNt_i).sum()
        )
    return ZZt_dp_i_

#
###
```

```python
############
## CHECKS ##
############

## Sample parameters
beta_sn0_i = 0.8
beta_sn1_i = 0.0
beta_sn2_i = 0.1
bbeta_dp0_i = [0.0, 0.02]
bbeta_dp1_i = [10.0, 20.0]
bbeta_dp2_i = [-0.1, -0.1]

## State variables
zz = torch.linspace(1,100,100)
NNt_i = 10 * pdf(f_normal(zz, 40, 3))
NNt = [30, 100]

## Compute quantities
Zt_d_i_ = Zt_d_i(zz, NNt_i, beta_sn0_i, beta_sn1_i, beta_sn2_i, NNt, bbeta_dp0_i, bbeta_dp1_i, bbeta_dp2_i)
Zt_dn_i_ = Zt_dn_i(zz, NNt_i, beta_sn0_i, beta_sn1_i, beta_sn2_i)
ZZt_dp_i_ = ZZt_dp_i(zz, NNt_i, beta_sn0_i, beta_sn1_i, beta_sn2_i, NNt, bbeta_dp0_i, bbeta_dp1_i, bbeta_dp2_i)

## Visualise
print(f"Dead biomass (all causes): {Zt_d_i_}")
print(f"Dead biomass (non-predation): {Zt_dn_i_}")
for i in range(len(NNt)):
  print(f"Dead biomass (predation by {i+1}): {ZZt_dp_i_[i]}")
print(f"Sum of dead biomass (predation by all): {Zt_dn_i_ + torch.tensor(ZZt_dp_i_).sum()}")

#
###
```

## Partitioning mortality of unstructured resources

Following a similar line of thought for grass, which is not structured by phenotype, we also partition its mortality into different causes.
Using the previous equations we arrive at

\begin{aligned}
G_{d}^{(0)}(t) &= G_{d,n}^{(0)}(t) + \sum_{j=1}^{J} G_{d,h}^{(0,j)}(t) \\
&= p_{d,n}^{(0)} \cdot G^{(0)}(t) + \sum_{j=1}^{J}  p_{s,n}^{(0)} \cdot p_{d,h}^{(0)} \cdot \frac{p_{d,h}^{(0,j)}}{\sum_{k=1}^{I} p_{d,h}^{(0,k)}} \cdot G^{(0)}(t) \\
\end{aligned}

where $G_d^{(0)}(t)$ is the biomass of dead grass, $G_{d,n}^{(0)}$ and $G_{d,h}^{(0,j)}$ is the biomass of grass that does not survive natural causes of death and herbivory by species $j$, respectively.
The probability of survival to herbivory is modelled as

\begin{equation}
p_{s,h}^{(0,j)} = \exp \left\{ N^{(j)}(t) \cdot \log \left\{ p_{s,h}^{(0,j,k)}\right\} \right\}, \phantom{\int}
\end{equation}

where the probability of survival to an encounter with a single grazer $p_{s,h}^{(0,j,k)}$ is a constant.



```python
#########################################
## PARTITIONING UNSTRUCTURED RESOURCES ##
#########################################

def p_sh_0j(Nt_j, beta_dh0_0j):
    p_sh_0j_ = torch.exp(Nt_j * torch.log(torch.tensor(1 - beta_dh0_0j)))
    return p_sh_0j_

def p_sh_0(NNt, bbeta_dh0_0):
    p_sh_0_ = 1.0
    for j in range(len(NNt)):
      p_sh_0_ = p_sh_0_ * p_sh_0j(NNt[j], bbeta_dh0_0[j])
    return p_sh_0_

def p_sn_0(beta_sn0_0):
    return beta_sn0_0

def p_s_0(beta_sn0_0, NNt, bbeta_dh0_0):
    p_sh_0_ = p_sh_0(NNt, bbeta_dh0_0)
    p_sn_0_ = p_sn_0(beta_sn0_0)
    p_s_0_ = p_sn_0_ * p_sh_0_
    return p_s_0_

def total_p_dh_0(NNt, bbeta_dh0_0):
    p_dh_0_ = 0.0
    for j in range(len(NNt)):
      p_dh_0_ = p_dh_0_ + (1 - p_sh_0j(NNt[j], bbeta_dh0_0[j]))
    return p_dh_0_

def Gt_d_0(Gt_0, beta_sn0_0, NNt, bbeta_dh0_0):
    Gt_d_0_ = (1 - p_s_0(beta_sn0_0, NNt, bbeta_dh0_0)) * Gt_0
    return Gt_d_0_

def Gt_dn_0(Gt_0, beta_sn0_0):
    Gt_dn_0_ = (1 - p_sn_0(beta_sn0_0)) * Gt_0
    return Gt_dn_0_

def GGt_dh_0(Gt_0, beta_sn0_0, NNt, bbeta_dh0_0):
    p_sh_0_ = p_sh_0(NNt, bbeta_dh0_0)
    total_p_dh_0_ = total_p_dh_0(NNt, bbeta_dh0_0)
    GGt_dh_0_ = []
    for j in range(len(NNt)):
      if total_p_dh_0_ == 0:
        GGt_dh_0_.append(0)
      else:
        GGt_dh_0_.append(
            (Gt_0 * p_sn_0(beta_sn0_0) * (1 - p_sh_0_) * (1 - p_sh_0j(NNt[j], bbeta_dh0_0[j])) / total_p_dh_0_)
        )
    return GGt_dh_0_

#
###
```

```python
############
## CHECKS ##
############

## Sample parameters
beta_sn0_0 = 0.8
bbeta_dh0_0 = [0.01, 0.02]

## State variables
NNt = [30, 100]
Gt_0 = 100

## Compute quantities
Gt_d_0_ = Gt_d_0(Gt_0, beta_sn0_0, NNt, bbeta_dh0_0)
Gt_dn_0_ = Gt_dn_0(Gt_0, beta_sn0_0)
GGt_dh_0_ = GGt_dh_0(Gt_0, NNt, bbeta_dh0_0)

## Check
print(f"Dead grass (all causes): {Gt_d_0_}")
print(f"Dead grass (non-herbivory): {Gt_dn_0_}")
for i in range(len(NNt)):
  print(f"Dead grass (herbivory by {i+1}): {GGt_dh_0_[i]}")
print(f"Total dead grass: {Gt_dn_0_ + torch.tensor(GGt_dh_0_).sum()}")

#
###
```

## Acquisition of biomass through herbivory, scavenging, and predation

Before the individuals can consume their food, they need to make sure it is dead. This can be achieved through herbivory, scavenging, or predation. This means that the amount of food available to a population is what it was able to kill or scavenge:

\begin{aligned}
Z_c^{(i)}(t) &= \lambda_1^{(i)} \cdot P_{d,h}^{(0,j)}(t) + \lambda_2^{(i)} \cdot G_{d,h}^{(0,i)}(t) + \lambda_3^{(i)} \cdot\sum_{j=1}^{J}  Z_{d,n}^{(j)}(t) + \lambda_4^{(i)} \cdot \sum_{j=1}^{J} Z_{d,p}^{(j,i)}(t), \phantom{\int} \\
\end{aligned}

where $Z_c^{(i)}$ is the total biomass available for consumption by the $i$-th species, $\lambda_1^{(i)}$ is the proportion of time spent grazing, $\lambda_2^{(i)}$ is the proportion of time spent scavenging, and $\lambda_3^{(i)}$ is the proportion of time spent consuming their kills.
These times are subjected to within-species constraints: individuals cannot do all three tasks simultaneously which implies $\sum_k \lambda_{k}^{(i)} \leq 1$.
For scavenging, species have to share biomass with other species that may also allocate time to scavenging, such that $\sum_i \lambda_2^{(i)} \leq 1$.
The time spent consuming their own kills (including herbivory) is already partitioned into species-specific components contributions.

Consequently, the amount of biomass that is not consumed is given by

\begin{aligned}
Z_{\neg c}(t) &= \sum_{i=1}^{I} \left(1 - \lambda_1^{(i)}\right) \cdot P_{d,h}^{(0,i)}(t) + \sum_{i=1}^{I} \left(1 - \lambda_2^{(i)}\right) \cdot G_{d,h}^{(0,i)}(t) + \left(1 - \sum_{i=1}^{I}\lambda_3^{(i)} \right) \cdot\sum_{j=1}^{J}  Z_{d,n}^{(j)}(t) + \\& \sum_{i=1}^{I} \left(1 - \lambda_4^{(i)}\right) \cdot \sum_{j=1}^{J} Z_{d,p}^{(j,i)}(t). \phantom{\int} \\
\end{aligned}


```python
############################
## ACQUISITION OF BIOMASS ##
############################

## Nutrients parameters
beta_sn0_P = 1.0
bbeta_dh0_P = [0.01, 0.02]

## Grazing parameters
beta_sn0_0 = 1.0
bbeta_dh0_0 = [0.01, 0.02]

## Survival parameters
bbeta_sn0 = [0.8, 0.8]
bbeta_sn1 = [0.0, 0.0]
bbeta_sn2 = [0.1, 0.1]
bbbeta_dp0 = [[0.0, 0.02], [0.02, 0.01]]
bbbeta_dp1 = [[10.0, 20.0], [10.0, 20.0]]
bbbeta_dp2 = [[-0.1, -0.1], [-0.1, -0.1]]

## Diet preference parameters
lllambda = [[0.0, 0.1, 0.2, 0.7],
            [0.0, 0.1, 0.5, 0.4]]

## State variables
zz = torch.linspace(1,100,100)
NNt = [30, 100]
NNNt = [NNt[0] * pdf(f_normal(zz, 40, 3)), NNt[1] * pdf(f_normal(zz, 40, 3))]
Pt_0 = torch.tensor(100)
Gt_0 = torch.tensor(100)

## Check
ZZt = [(zz * NNt[0] * pdf(f_normal(zz, 40, 3))).sum(), (zz * NNt[1] * pdf(f_normal(zz, 40, 3))).sum()]

## Compute quantities of nutrients consumed
Pt_d_ = Gt_d_0(Pt_0, beta_sn0_P, NNt, bbeta_dh0_P)
Pt_dn_ = Gt_dn_0(Pt_0, beta_sn0_P)
PPt_dh_ = GGt_dh_0(Pt_0, beta_sn0_P, NNt, bbeta_dh0_P)

## Check
print(Pt_d_)
print(Pt_dn_)
print(PPt_dh_)
print(Pt_dn_ + torch.tensor(PPt_dh_).sum())

## Compute quantities of grass consumed
Gt_d_ = Gt_d_0(Gt_0, beta_sn0_0, NNt, bbeta_dh0_0)
Gt_dn_ = Gt_dn_0(Gt_0, beta_sn0_0)
GGt_dh_ = GGt_dh_0(Gt_0, beta_sn0_0, NNt, bbeta_dh0_0)

## Check
print(Gt_d_)
print(Gt_dn_)
print(GGt_dh_)
print(Gt_dn_ + torch.tensor(GGt_dh_).sum())

## For each predator
ZZt_d = []
ZZt_dn = []
ZZZt_dp = []
for i in range(len(NNt)):

  ## Compute dead biomass
  ZZt_d.append(Zt_d_i(zz, NNNt[i], bbeta_sn0[i], bbeta_sn1[i], bbeta_sn2[i], NNt, bbbeta_dp0[i], bbbeta_dp1[i], bbbeta_dp2[i]))
  ZZt_dn.append(Zt_dn_i(zz, NNNt[i], bbeta_sn0[i], bbeta_sn1[i], bbeta_sn2[i]))
  ZZZt_dp.append(ZZt_dp_i(zz, NNNt[i], bbeta_sn0[i], bbeta_sn1[i], bbeta_sn2[i], NNt, bbbeta_dp0[i], bbbeta_dp1[i], bbbeta_dp2[i]))

## Check
print(ZZt_d)
print(ZZt_dn)
print(ZZZt_dp)
print(ZZt_dn[0] + torch.tensor(ZZZt_dp[0]).sum())
print(ZZt_dn[1] + torch.tensor(ZZZt_dp[1]).sum())

## Available biomass through scavenging and predation
Zt_dn_ = torch.tensor(ZZt_dn).sum(0)
ZZt_dp_ = torch.tensor(ZZZt_dp).sum(1)

## Check
print(Zt_dn_)
print(ZZt_dp_)

## Compute acquired biomass
ZZt_c = []
for i in range(len(NNt)):
  ZZt_c.append(lllambda[i][0] * PPt_dh_[i] + lllambda[i][1] * GGt_dh_[i] + lllambda[i][2] * Zt_dn_ + lllambda[i][3] * ZZt_dp_[i])

## Compute not acquired biomass
ZZt_notc = (1 - torch.tensor(lllambda)[:,2].sum()) * Zt_dn_
for i in range(len(NNt)):
  ZZt_notc += (1 - lllambda[i][0]) * PPt_dh_[i] + (1 - lllambda[i][1]) * GGt_dh_[i] + (1 - lllambda[i][3]) * ZZt_dp_[i]

## Check
print(ZZt_c)
print(ZZt_notc)
print(torch.tensor(ZZt_c).sum() + ZZt_notc)
print(torch.tensor(PPt_dh_).sum() + torch.tensor(GGt_dh_).sum() + Zt_dn_ + ZZt_dp_.sum())

#
###
```

## Consumption

To avoid potential over-consumption of other ecosystem compartments (i.e. consuming an amount of biomass that exceeds the available amount of biomass), we choose a Michaelis-Menten curve to model the total biomass of grass consumed by a single individual in a year:

\begin{equation}
c^{(i)}(z,t) = \frac{\gamma_{1}^{(i)} \gamma_{2}^{(i)} Z_c^{(i)}(t) }{\gamma_{1}^{(i)} N^{(i)}(t) + \gamma_{2}^{(i)} Z_c^{(i)}(t)},
\end{equation}

where $\gamma_{1}^{(i)}$ is the per-capita feeding rate of species $i$ (e.g. in tons of biomass per year), and $\gamma_{2}^{(i)}$ is the maximum fraction of the total amount of biomass available that can be consumed by species $i$, and $Z_c^{(i)}(t)$ denotes the total amount of biomass available for consumption and assimilation.

```python
######################
## CONSUMPTION RATE ##
######################

## Define the Michaelis-Menten consumption function
def consumption(z_c, gamma_1, gamma_2, N_i):
    return (gamma_1 * gamma_2 * z_c) / (gamma_1 * N_i + gamma_2 * z_c)

## Define parameter sets for visualization
param_sets = [
    {"gamma_1": 0.00144, "gamma_2": 0.8, "N_i": 1, "label": "Aspen"},
    {"gamma_1": 6.81, "gamma_2": 0.5, "N_i": 1, "label": "Elk"},
    {"gamma_1": 13.33, "gamma_2": 0.5, "N_i": 1, "label": "Bison"},
    {"gamma_1": 6.7, "gamma_2": 0.68, "N_i": 1, "label": "Wolves"},
    {"gamma_1": 12.0, "gamma_2": 0.68, "N_i": 1, "label": "Cougars"},
]

## Range of available biomass values (converted to PyTorch tensor)
z_c_values = torch.linspace(0, 20, 1000)  # Available biomass from 0 to 100 tons

## Plot consumption functions for different parameter sets
plt.figure(figsize=(6, 4))
for params in param_sets:
    consumption_values = consumption(z_c_values, params["gamma_1"], params["gamma_2"], params["N_i"])
    plt.plot(z_c_values.numpy(), consumption_values.numpy(), label=params["label"])
plt.plot(z_c_values.numpy(), z_c_values.numpy(), 'k--', label="1:1 Line")

## Formatting the plot
plt.xlabel("Available Biomass $Z_c^{(i)}(t)$ (kg)")
plt.ylabel("P.c. Consumption Rate $c^{(i)}(z,t)$ (kg.day$^{-1}$)")
# plt.title("Michaelis-Menten Consumption as a Function of Available Biomass")
plt.legend(fontsize=8, loc="upper left", bbox_to_anchor=(1, 1))
# plt.grid()

## Show the plot
plt.show()

#
###
```


## Assimilation

We also introduce constraints on assimilation and conversion of the consumed biomass:

\begin{equation}
g^{(i)}(z,t) = \alpha_1^{(i)} \alpha_2^{(i)} c^{(i)}(z, t),
\end{equation}

where $\alpha_1^{(i)}$ is the proportion of the consumed biomass that is assimilated by species $i$, and $\alpha_2^{(i)}$ is the conversion efficiency of consumed biomass into biomass of species $i$, which relates to the quality (i.e. nutrient content) of the food consumed.



```python
##################################
## CONSUMPTION AND ASSIMILATION ##
##################################

def g_i(z, alpha_1_i, alpha_2_i, gamma_1_i, Nt_i, gamma_2_i, Zt_c_i):
    g_i_ = alpha_1_i * alpha_2_i * c_i(z, gamma_1_i, Nt_i, gamma_2_i, Zt_c_i)
    return g_i_ * torch.ones_like(z)

def c_i(z, gamma_1_i, Nt_i, gamma_2_i, Zt_c_i):
    c_i_ = gamma_1_i * gamma_2_i * Zt_c_i / (gamma_1_i * Nt_i + gamma_2_i * Zt_c_i)
    return c_i_ * torch.ones_like(z)

#
###
```

```python
############
## CHECKS ##
############

## Generate data
xx = torch.linspace(0, 10, 100)  # Arbitrary range for visualization

## Sample parameters
alpha_1_i = 0.8
alpha_2_i = 0.24
gamma_1_i = 6.7
gamma_2_i = 0.68
Nt_i = 1.0
Zt_c_i = 0.0

## Compute function values
c_values = c_i(torch.tensor(0), gamma_1_i, Nt_i, gamma_2_i, xx)
g_values = g_i(torch.tensor(0), alpha_1_i, alpha_2_i, gamma_1_i, Nt_i, gamma_2_i, xx)

## Plot results
plt.figure(figsize=(10, 4))
#
plt.subplot(1, 2, 1)
plt.plot(xx, c_values, label=r'$c_i(z)$', color='blue')
plt.plot(xx, xx, '--')
plt.xlabel('z')
plt.ylabel('c_i')
plt.title('c_i(z) vs z')
plt.legend()
#
plt.subplot(1, 2, 2)
plt.plot(xx, g_values, label=r'$g_i(z)$', color='red')
plt.plot(xx, xx, '--')
plt.xlabel('z')
plt.ylabel('g_i')
plt.title('g_i(z) vs z')
plt.legend()
#
plt.tight_layout()
plt.show()

#
###
```

## Maintenance

Individuals have to pay maintenance costs due to metabolic activity.
For the sake of simplicity we assume that these costs are proportional to the square of the body weight of an individual:

\begin{equation}
l_{m}^{(i)}(z, t) = \delta^{(i)} \cdot z^{3/4}.
\end{equation}

```python
####################
## METABOLIC RATE ##
####################

## Define function
def metabolic_cost(z, delta):
    return delta * z**1.25

## Define parameter sets for visualization
param_sets = []
for delta in [0.01, 0.1, 0.2, 0.3]:
  param_sets.append({
      "delta": delta,
      "label": rf"$\delta={delta}$"
  })

## Range of biomass values (converted to PyTorch tensor)
z_values = torch.linspace(0, 100, 1000)  # Available biomass from 0 to 100 tons

## Plot consumption functions for different parameter sets
plt.figure(figsize=(6, 4))
for params in param_sets:
    metabolic_cost_values = metabolic_cost(z_values, params["delta"])
    plt.plot(z_values.numpy(), metabolic_cost_values.numpy(), label=params["label"])
plt.plot(z_values.numpy(), z_values.numpy(), 'k--', label="1:1 Line")

## Formatting the plot
plt.xlabel("Body Mass $z$ (kg)")
plt.ylabel("Metabolic Cost $l_m^{(i)}(z,t)$ (kg.day$^{-1}$)")
plt.legend(fontsize=8, loc="upper left", bbox_to_anchor=(1, 1))
# plt.grid()

## Show the plot
plt.show()

#
###
```

## Reproduction
The probability of breeding, $p_{b}(z)$, is defined as a sigmoid function of body size,

\begin{equation}
p_b^{(i)}(z) = \beta_{b,0}^{(i)} \cdot \mathrm{sigmoid} \left\{\left( z - \beta_{b,1}^{(i)} \right) \beta_{b,2}^{(i)} \right\},
\end{equation}

where $\beta_{b,0}^{(i)} \in [0,1]$ is the maximum breeding probability, $\beta_{b,1}^{(i)} \in \mathbb{R}^{+}$ is the body mass at breeding onset, and $\beta_{b,2}^{(i)}$ is the increase in log-odds of breeding per unit of body mass.
For the sake of simplicity, we assume that the number of offspring produced by breeders is determined by a fixed fraction of body mass that the individuals invest in reproduction:

\begin{equation}
n_{o}^{(i)}(z) = \nu_{o}^{(i)} \cdot \frac{z}{\mu_{o}^{(i)}},
\end{equation}

where $\nu_{o}^{(i)}$ designates the fraction of body reserves allocated to offspring production by the individuals of species $i$. This allows us to define the loss term for reproduction:

\begin{equation}
l_{r}^{(i)}(z,t) = \mu_{o}^{(i)} \cdot p_{b}^{(i)}(z) \cdot n_{o}^{(i)}(z),
\end{equation}

which means that the upper bound of the total loss in body mass that occurs through reproduction is $\int l_{r}(z,t) N(z,t) dz \leq \nu_{o} \bar{z} N(t) \leq \bar{z} N(t)$.
In other words, individuals cannot lose more than what they have.

The phenotype of offspring is then determined according to the Gaussian model

\begin{aligned}
& f_{r}^{(i)}(z | z') = \frac{1}{\sqrt{2 \pi {\sigma_o^{(i)}}^2}} \exp \left \{ -\frac{1}{2 {\sigma_o^{(i)}}^2} \left(z - \mu_o^{(i)} \right)^2 \right \} \phantom{\int} \\
\end{aligned}

where $\mu_o^{(i)}$ and $\sigma_o^{(i)}$ the mean and variance of the body mass of offspring of species $i$.

```python
##########################
## BREEDING PROBABILITY ##
##########################

## Define function using PyTorch
def breeding_probability_i(z, beta_0, beta_1, beta_2):
    return beta_0 * torch.sigmoid((z - beta_1) * beta_2)

## Define parameter sets for visualization
param_sets = [
    {"beta_0": 1.0, "beta_1": 2.4, "beta_2": 1.0, "label": "Aspen"},
    {"beta_0": 1.0, "beta_1": 300.0, "beta_2": 0.025, "label": "Elk"},
    {"beta_0": 0.9, "beta_1": 400.0, "beta_2": 0.025, "label": "Bison"},
    {"beta_0": 1.0, "beta_1": 40.0, "beta_2": 0.11, "label": "Wolves"},
    {"beta_0": 1.0, "beta_1": 40.0, "beta_2": 0.1, "label": "Cougars"},
]
## Range of body mass values (converted to PyTorch tensor)
z_values = torch.linspace(1, 1000, 1000)

## Plot survival functions for different parameter sets
plt.figure(figsize=(8, 6))
for params in param_sets:
    y = breeding_probability_i(z_values, params["beta_0"], params["beta_1"], params["beta_2"])
    plt.plot(z_values.numpy(), y.numpy(), label=params["label"])

## Formatting the plot
plt.xlabel("Body Mass (z) (kg)")
plt.ylabel("Breeding Probability")
# plt.title("Intrinsic Survival as a Function of Body Mass")
plt.legend(fontsize=8, loc="upper left", bbox_to_anchor=(1, 1))
plt.xscale("log")
plt.grid()

## Show the plot
plt.show()

#
###
```

```python
##################################
## REPRODUCTION AND MAINTENANCE ##
##################################

def l_i(z, nu_o_i, mu_o_i, beta_b0_i, beta_b1_i, beta_b2_i, delta_i):
    l_i_ =  l_m_i(z, delta_i) + l_r_i(z, nu_o_i, mu_o_i, beta_b0_i, beta_b1_i, beta_b2_i)
    return l_i_

def l_m_i(z, delta_i):
    l_m_i_ = delta_i * z**1.25
    return l_m_i_

def l_r_i(z, nu_o_i, mu_o_i, beta_b0_i, beta_b1_i, beta_b2_i):
    l_r_i_ = mu_o_i * p_b_i(z, beta_b0_i, beta_b1_i, beta_b2_i) * n_o_i(z, nu_o_i, mu_o_i)
    return l_r_i_

def p_b_i(z, beta_b0_i, beta_b1_i, beta_b2_i):
    p_b_i_ = beta_b0_i * torch.sigmoid((z - beta_b1_i) * beta_b2_i)
    return p_b_i_

def n_o_i(z, nu_o_i, mu_o_i):
    n_o_i_ = nu_o_i * z / mu_o_i
    # n_o_i_ = torch.tensor(1)
    return n_o_i_

#
###
```

## Growth

We established that the total biomass gained by an individual through consumption and assimilation is $g^{(i)}(z,t)$.
The total biomass lost by an individual with phenotype $z$ is obtained by summing the reproduction and maintenance costs, $l^{(i)}(z,t) = l_r^{(i)}(z,t) + l_m^{(i)}(z,t)$.
We assume that the expectation of the body mass of individuals that survive between $t$ and $t+1$ can be obtained by computing the balance of gains and losses following

\begin{aligned}
& \mu_{s}^{(i)}(z') = z' + g^{(i)}(z', t) - l^{(i)}(z',t). \phantom{\int} \\
\end{aligned}

We assume that the probability density functions (pdfs) of the phenotype of survivors is a normal distribution around the expectation of body mass change between $t$ and $t+1$,

\begin{aligned}
& f_{s}^{(i)}(z | z') = \frac{1}{\sqrt{2 \pi {\sigma_s^{(i)}}^2}} \exp \left \{ -\frac{1}{2 {\sigma_s^{(i)}}^2} \left(z - \mu_s^{(i)}(z') \right)^2 \right \}, \phantom{\int} \\
\end{aligned}

where the mean body mass of survivors depends on the balance between biomass gained through consumption and lost through reproduction and maintenance.


```python
#############
## KERNELS ##
#############

def f_s_i(z, zp, sigma_s_i, alpha_1_i, alpha_2_i, gamma_1_i, Nt_i, gamma_2_i, Zt_c_i, nu_o_i, mu_o_i, beta_b0_i, beta_b1_i, beta_b2_i, delta_i):
    f_s_i_ = f_normal(z, mu_s_i(zp, alpha_1_i, alpha_2_i, gamma_1_i, Nt_i, gamma_2_i, Zt_c_i, nu_o_i, mu_o_i, beta_b0_i, beta_b1_i, beta_b2_i, delta_i), sigma_s_i)
    return f_s_i_

def mu_s_i(z, alpha_1_i, alpha_2_i, gamma_1_i, Nt_i, gamma_2_i, Zt_c_i, nu_o_i, mu_o_i, beta_b0_i, beta_b1_i, beta_b2_i, delta_i):
    mu_s_i_ = z + g_i(z, alpha_1_i, alpha_2_i, gamma_1_i, Nt_i, gamma_2_i, Zt_c_i) - l_i(z, nu_o_i, mu_o_i, beta_b0_i, beta_b1_i, beta_b2_i, delta_i)
    return mu_s_i_

def f_r_i(z, zp, mu_o_i, sigma_o_i):
    f_r_i_ = f_normal(z, mu_o_i, sigma_o_i)
    return f_r_i_

#
###
```

## Excretion

A key component of this model is the recycling of biomass through decomposition.
In order to work out how much biomass is recycled, we need to compute waste products from (1) herbivory, scavenging, and predation, as well as (2) those coming from imperfect consumption and assimilation. The amount of biomass excreted is considered to be all biomass that is not ingested nor assimilated or that is used for metabolic maintenance

\begin{equation}
Z_{e}(t) = \sum_{i=1}^{I} Z_c^{(i)}(t) - \int g^{(i)}(z', t) p_s^{(i)}(z') N^{(i)}(z', t) dz' + \int l_{m}^{(i)}(z',t) p_s^{(i)}(z') N^{(i)}(z', t) dz'.
\end{equation}


```python
###############
## EXCRETION ##
###############

def Zt_e_i(zz, NNt_i, alpha_1_i, alpha_2_i, gamma_1_i, Nt_i, gamma_2_i, Zt_c_i, delta_i, beta_sn0_i, beta_sn1_i, beta_sn2_i, NNt, bbeta_dp0_i, bbeta_dp1_i, bbeta_dp2_i):
    Zt_e_i_ = Zt_c_i - ((g_i(zz, alpha_1_i, alpha_2_i, gamma_1_i, Nt_i, gamma_2_i, Zt_c_i) - l_m_i(zz, delta_i)) * p_s_i(zz, beta_sn0_i, beta_sn1_i, beta_sn2_i, NNt, bbeta_dp0_i, bbeta_dp1_i, bbeta_dp2_i) * NNt_i).sum()
    return Zt_e_i_

#
###
```

## Full ecosystem dynamics

The updated ecosystem dynamics, now accounting for structured species, are given below:

\begin{aligned}
\Delta D(t) &= \phi_{BD} B(t) + \phi_{GD} G(t) - hh(\phi_{DB} D(t), \rho_B B(t)) + Z_e(t) + Z_{\neg c}(t) \phantom{\int} \\
\Delta B(t) &= (1 - \phi_{DP}) \cdot hh(\phi_{DB} D(t), \rho_B B(t)) - \phi_{BD} B(t)\phantom{\int} \\
\Delta P(t) &= \phi_{DP} \cdot hh(\phi_{DB} D(t), \rho_B B(t)) - hh(\phi_{PG} P(t), \rho_G G(t)) - P_c(t) \phantom{\int} \\
\Delta G(t) &= hh(\phi_{PG} P(t), \rho_G G(t)) - \phi_{GD} G(t) - G_c(t). \phantom{\int} \\
\end{aligned}

where the dynamics of the dead organic matter $D$ are now impacted by excretion $Z_e(t)$ and waste products of consumption $Z_{\neg c}$, and the dynamics of nutrients and grass are impacted by consumption of nutrients $P_c(t)$ and grass $G_c(t)$.

# Supplementary

## From IPM to difference equation

We can re-write the IPM as a difference equation as follows:

\begin{aligned}
N(z, t + 1) &= \int_{S_z} \left \{ f_r(z | z') p_{b}(z') p_{r} n_{o}(z') + f_s(z | z') \right \} p_s(z') N(z', t) dz' \\
&= \int f_r(z | z') p_{b}(z') p_{r} n_{o}(z') p_s(z') N(z', t) dz' + \int f_s(z | z') p_s(z') N(z', t) dz' \\
&= \int f_r(z | z') p_{b}(z') p_{r} n_{o}(z') p_s(z') N(z', t) dz' + \int f_s(z | z') N(z', t) dz' - \int f_s(z | z') p_d(z') N(z', t) dz' \\
\Leftrightarrow \Delta N(z, t) &= \int f_r(z | z') p_{b}(z') p_{r} n_{o}(z') p_s(z') N(z', t) dz' + \int f_s(z | z') N(z', t) dz' - \int f_s(z' | z) N(z, t) dz' - \int f_s(z | z') p_d(z') N(z', t) dz' \\
\Rightarrow \int \Delta N(z, t) dz &= \iint f_r(z | z') p_{b}(z') p_{r} n_{o}(z') p_s(z') N(z', t) dz' dz - \iint f_s(z | z') p_d(z') N(z', t) dz' dz \\
&= \int p_{b}(z') p_{r} n_{o}(z') p_s(z') N(z', t) dz' - \int p_d(z') N(z', t) dz'. \\
\end{aligned}


## Verification of the constraints



We hence investigate the total change in population density and biomass between $t$ and $t+1$:

\begin{aligned}
&\Delta N(t) = N(t+1) - N(t) = \int N(z,t+1) dz - \int N(z,t) dz \\
&\Delta Z(t) = Z(t+1) - Z(t) = \int z N(z,t+1) dz - \int z N(z,t) dz. \\
\end{aligned}



We first consider the total change in population density:

\begin{aligned}
\Delta N(t) &= \iint f_r(z | z') p_{b}(z') p_{r} n_{o}(z') p_{s}(z') N(z', t) dz' dz + \iint f_s(z | z') p_s(z') N(z', t) dz' dz - \int N(z,t) dz \\
&= \int p_{b}(z') p_{r} n_{o}(z') p_{s}(z') N(z', t) dz' + \int p_s(z') N(z', t) dz' - \int N(z,t) dz \\
&\leq \int n_{o}(z') N(z', t) dz' + \int N(z', t) dz' - \int N(z,t) dz, ~(\mathrm{assuming}~ p_s(z') = 1 ~\forall~ z') \\
&= N_{o}(t), \phantom{\int} \\
\end{aligned}

which demonstrates that the total population growth is bounded by total offspring production.



Similarly, for the total change in the population biomass we get:

\begin{aligned}
\Delta Z(t) &= \int z \int f_r(z | z') p_{b}(z') p_{r} n_{o}(z') p_{s}(z') N(z', t) dz' dz + \int z \int f_s(z | z') p_s(z') N(z', t) dz' dz - \int z N(z,t) dz \\
&= \int \mathrm{E}_{f_r}[z|z'] p_{b}(z') p_{r} n_{o}(z') p_{s}(z') N(z', t) dz' + \int \mathrm{E}_{f_s}[z|z'] f_s(z | z') p_s(z') N(z', t) dz' - \int z N(z,t) dz. \\
\end{aligned}

To go further, we need to assume specific functions for the probability density functions of reproduction and survival.



We assume that both the reproduction and survival kernels are normal probability density functions:
Namely that $z \sim \mathcal{N}(z_o, \sigma_o)$, for the body mass of offsrping, and that $z \sim \mathcal{N}(z' + \Delta z, \sigma_s)$, for the body mass of survivors.

\begin{aligned}
\Delta Z(t) &= \int z_o p_{b}(z') p_{r} n_{o}(z') p_{s}(z') N(z', t) dz' + \int (z' + \Delta z~) p_s(z') N(z', t) dz' - \int z N(z,t) dz. \\
\end{aligned}


At this point we need to introduce the dynamics of the phenotype of the individuals:

\begin{aligned}
\Delta z(z, t) &= g(z, t) - l(z, t) = g(z, t) - z_{o} p_{b}(z) n_{o}(z),
\end{aligned}

where the function $g(z,t)$ and $l(z,t)$ correspond to the gain and loss of biomass, respectively, experienced by the individuals.
Here we assume that the fraction of breeding individuals, $p_{b}(z)$, lose an amount of body mass equal to the biomass of offspring produced, $z_{o} n_{o}(z)$.

This implies that

\begin{aligned}
\Delta Z(t) &= \int z_{o} p_{b}(z') p_{r} n_{o}(z') p_{s}(z') N(z', t) dz' + \int z' p_s(z') N(z', t) dz' + \int g(z',t) p_s(z') N(z',t) dz' - \int z_{o} p_{b}(z') n_{o}(z') p_s(z') N(z',t) dz' - \int z N(z, t) dz \\
&= (p_{r} - 1) \int z_{o} p_{b}(z') n_{o}(z') p_{s}(z') N(z', t) dz' + \int z' p_s(z') N(z', t) dz' + \int g(z', t) p_s(z') N(z',t) dz' - \int z N(z, t) dz \\
&\leq \int z' p_s(z') N(z', t) dz' + \int g(z', t) p_s(z') N(z',t) dz' - \int z N(z, t) dz, ~(\mathrm{assuming}~ p_r = 1) \\
&\leq \int g(z', t) N(z',t) dz', ~(\mathrm{assuming}~ p_s(z') = 1 ~\forall~ z') \\
&= \bar{g}(t) N(t) \phantom{\int}
\end{aligned}

This means that no biomass is lost if the recruitment of the offspring is $100\%$, namely when $p_{r} = 1$.
Then an increase in total population biomass, $Z(t)$, is only possible through the gains function $g(z, t)$.

## Impact of multiple predators on elk population dynamics

In order to partition the amount of biomass available to each predator we compute the impact of each predator on the population of elk:

\begin{aligned}
p_s(z) &= p_{s,n}(z) \cdot p_{s,p}(z) \\
&= p_{s,n}(z) \cdot p_{s,pw}(z) \cdot p_{s,pm}(z) \\
&= p_{s,n}(z) \cdot \left( 1 - p_{d,pw}(z) \right) \cdot \left( 1 - p_{d,pm}(z) \right) \\
&= p_{s,n}(z) - p_{s,n}(z) \cdot p_{d,pw}(z) - p_{s,n}(z) \cdot p_{d,pm}(z) + p_{s,n}(z) \cdot p_{d,pw}(z) \cdot p_{d,pm}(z) \\
\Leftrightarrow p_d(z) &= 1 - p_{s,n}(z) + p_{s,n}(z) \cdot p_{d,pw}(z) + p_{s,n}(z) \cdot p_{d,pm}(z) - p_{s,n}(z) \cdot p_{d,pw}(z) \cdot p_{d,pm}(z) \\
&= p_{d,n}(z) + p_{s,n}(z) \cdot p_{d,pw}(z) + p_{s,n}(z) \cdot p_{d,pm}(z) - p_{s,n}(z) \cdot p_{d,pw}(z) \cdot p_{d,pm}(z) \\
&= p_{d,n}(z) + p_{s,n}(z) \cdot p_{d,pw}(z) + p_{s,n}(z) \cdot p_{s,pw}(z) \cdot p_{d,pm}(z) \\
\end{aligned}


## Modelling predation rates

I propose to model the number of elk killed every year by wolf as a Binomial distribution.
We build the model for predation by considering a single wolf:

\begin{aligned}
& N_{d}^{(1)} ~|~ N^{(0)} \sim \mathcal{B} \left( p_d, N^{(0)} \right) \phantom{\int} \\
& N_{d}^{(2)} ~|~ N^{(0)}, N_{d}^{(1)} \sim \mathcal{B} \left( p_d, N^{(0)} - N_{d}^{(1)} \right) \phantom{\int} \\
& \vdots \phantom{\int} \\
& N_{d}^{(i)} ~|~ N^{(0)}, N_{d}^{(1)}, \dots, N_{d}^{(i-1)} \sim \mathcal{B} \left( p_d, N^{(0)} - \sum_{j=1}^{i-1} N_{d}^{(j)} \right) \phantom{\int} \\
\end{aligned}

We break down this by considering first the individual expectations:

\begin{aligned}
& \mathrm{E} \left[ N_d^{(1)} \right] = p_d N^{(0)} \phantom{\int} \\
& \mathrm{E} \left[ N_d^{(2)} \right] = p_d \left( N^{(0)} - \mathrm{E} \left[ N_d^{(1)} \right] \right) = p_d \left( N^{(0)} - p_d N^{(0)} \right) \phantom{\int} \\
& \mathrm{E} \left[ N_d^{(3)} \right] = p_d \left( N^{(0)} - \mathrm{E} \left[ N_d^{(1)} \right] - \mathrm{E} \left[ N_d^{(2)} \right] \right) = p_d \left( N^{(0)} - p_d N^{(0)} - p_d \left( N^{(0)} - p_d N^{(0)} \right) \right) \phantom{\int} \\
\end{aligned}

We note that these expressions can be refactored as follows:

\begin{aligned}
& \mathrm{E} \left[ N_d^{(1)} \right] = N^{(0)} p_d \phantom{\int} \\
& \mathrm{E} \left[ N_d^{(2)} \right] = N^{(0)} p_d \left( 1 - p_d \right) \phantom{\int} \\
& \mathrm{E} \left[ N_d^{(3)} \right] = N^{(0)} p_d (1 - p_d - p_d (1 - p_d)) = N^{(0)} p_d (1 - p_d) (1 - p_d)  \phantom{\int} \\
\end{aligned}

This leads to the general form for the expectation:

\begin{aligned}
\mathrm{E} \left[ N_d^{(i)} \right] &= N^{(0)} p_d (1 - p_d)^{i-1} \phantom{\int} \\
\end{aligned}

We can hence compute the expected number of elk killed by the entire wolf population by evaluated all the conditional expectations:

\begin{aligned}
\mathrm{E} \left[ \sum_{i=1}^{W} N_d^{(i)} \right] &= \sum_{i=1}^{W} \mathrm{E} \left[ N_d^{(i)} \right] \phantom{\int} \\
&= \sum_{i=1}^{W} N^{(0)} p_d (1 - p_d)^{i-1} \phantom{\int} \\
&= N^{(0)} p_d \sum_{i=1}^{W} p_s^{i-1} \phantom{\int} \\
&= N^{(0)} p_d \sum_{i=1}^{W} p_s^{i-1} \phantom{\int} \\
\end{aligned}

For now we focus on the geometric series:

\begin{aligned}
\sum_{i=1}^{W} p_s^{i-1} &= 1 + p_s + p_s^{2} + \ldots + p_s^{W-1} \phantom{\int} \\
&= \left( 1 + p_s + p_s^{2} + \ldots + p_s^{W-1} \right) \frac{(1 - p_s)}{(1 - p_s)} \phantom{\int} \\
&= \left(\left(1 - p_s\right) + \left(p_s - p_s^{2}\right) + \left(p_s^{2} - p_s^{3}\right) + \ldots + \left(p_s^{W-1} - p_s^{W}\right) \right) (1 - p_s)^{-1} \phantom{\int} \\
&= \frac{1 - p_s^{W}}{1 - p_s} \phantom{\int} \\
\end{aligned}

Putting this together with the expectation of the total number of elk killed we get the expression:

\begin{equation}
\mathrm{E} \left[ \sum_{i=1}^{W} N_d^{(i)} \right] = N^{(0)}\left(1 - p_s^{W}\right)
\end{equation}

For stabler computations we can use the following formula for computing the proportion of the population killed by wolves each year:

\begin{equation}
p_{d,p} = 1 - \exp \left\{ W \log \left\{ p_s \right\} \right\}
\end{equation}
