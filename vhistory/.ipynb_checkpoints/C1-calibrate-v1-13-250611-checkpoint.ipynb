```python
#####################
## STANDARDISATION ##
#####################

def compute_nan_mean_std(x, dim=None, keepdim=False, eps=1e-8):
    mask = ~torch.isnan(x)
    count = mask.sum(dim=dim, keepdim=True)

    x_masked = torch.where(mask, x, torch.tensor(0.0, dtype=x.dtype, device=x.device))
    mean = x_masked.sum(dim=dim, keepdim=True) / count

    var = ((torch.where(mask, x - mean, torch.tensor(0.0, dtype=x.dtype, device=x.device))) ** 2).sum(dim=dim, keepdim=True) / count
    std = torch.sqrt(var + eps)

    if not keepdim and dim is not None:
        mean = mean.squeeze(dim)
        std = std.squeeze(dim)

    return mean, std

def apply_standardization(x, mean, std):
    z = (x - mean) / std
    z = torch.where(torch.isnan(x), x, z)  # Preserve original NaNs
    return z

def reverse_standardization(z, mean, std):
    x = z * std + mean
    x = torch.where(torch.isnan(z), z, x)  # Preserve original NaNs
    return x

MEAN, STD = compute_nan_mean_std(time_series_tensor, 0)
time_series_tensor_standardised = apply_standardization(time_series_tensor, MEAN, STD)

## Check
plt.plot(time_series_tensor_standardised)
# plt.yscale('log')
plt.show()

#
###
```

```python
########################
## FORMAT PREDICTIONS ##
########################

def format_predicted_time_series(predictions):
    zzz, tt, DDt, BBt, PPt, GGt, NNNt, NNNNzt, ttotal_biomass = predictions    
    nnnnzt = NNNNzt/(NNNNzt.sum(2)[:,:,None])
    time_series_tensor_predicted = torch.cat((
            GGt[:,None], # (num_time_steps) --> (num_time_steps, 1)
            (zzz * nnnnzt * MASK[:,None]).sum(2)[:,0:2]/100.0, # Trees upscaled by 100.0 kg --> downscaled to individuals
            NNNt[:,2:6] # Counts available only for animals 
        ), dim=1)
    return time_series_tensor_predicted

## Check
time_series_tensor_predicted = format_predicted_time_series(predictions)
print(time_series_tensor_predicted.shape)
# print(time_series_tensor_predicted)

#
###
```

```python
####################
## DEFINE WRAPPER ##
####################

## Model
model_r = YellowstoneEcosystemModel("parameters-v1-13.yaml")
model_p = YellowstoneEcosystemModel("parameters-v1-13.yaml")

## Probabilities to rates
def r2p(x):
    return 1 - 1/(x+1)
def p2r(x):
    return x/(1-x)

## Define wrapper
def wrapper(ttheta, t_max=30, model_r=model_r, model_p=model_p, Dt0=Dt0, Bt0=Bt0, Pt0=Pt0, Gt0=Gt0, NNt0=NNt0, NNNzt0=NNNzt0):
    
    ## Initiate idx
    idx = 0
    
    ## Initial conditions unstructured states
    Dt0_p = Dt0 * ttheta[idx]; idx += 1
    Bt0_p = Bt0 * ttheta[idx]; idx += 1
    Pt0_p = Pt0 * ttheta[idx]; idx += 1
    Gt0_p = Gt0 * ttheta[idx]; idx += 1
    
    ## Initial conditions structured states
    NNt0_p = NNt0.clone()
    NNNzt0_p = NNNzt0.clone()
    scaling = [1.0, 1.0, 1.3, 1.5, 1.0, 0.2]
    for i in range(len(NNt0_p)):
        NNt0_p[i] = NNt0[i] * ttheta[idx] * scaling[i]
        NNNzt0_p[i] = NNNzt0[i] * ttheta[idx] * scaling[i]
        idx += 1

    """
    ## Parameters (scaling by copying from model_r)
    for i in range(len(NNt0_p)):

        ## Nutrient assimilation and grazing (I * 2)
        model_p.bbeta_dh0_P[i] = r2p(p2r(model_r.bbeta_dh0_P[i]) * ttheta[idx]); idx += 1 # [0,1]
        model_p.bbeta_dh0_G[i] = r2p(p2r(model_r.bbeta_dh0_G[i]) * ttheta[idx]); idx += 1 # [0,1]

        ## Predation and scavenging (I * I * 4)
        for j in range(len(NNt0_p)):
            model_p.bbbeta_dp0[i][j] = r2p(p2r(model_r.bbbeta_dp0[i][j]) * ttheta[idx]); idx += 1 # [0,1]
            model_p.bbbeta_dp1[i][j] = model_r.bbbeta_dp1[i][j] * ttheta[idx]; idx += 1
            model_p.bbbeta_dp2[i][j] = model_r.bbbeta_dp2[i][j] * ttheta[idx]; idx += 1        
            model_p.bbbeta_dn0[i][j] = r2p(p2r(model_r.bbbeta_dn0[i][j]) * ttheta[idx]); idx += 1 # [0,1]

        ## Natural survival (I * 3)
        model_p.bbeta_sn0[i] = r2p(p2r(model_r.bbeta_sn0[i]) * ttheta[idx]); idx += 1 # [0,1]
        model_p.bbeta_sn1[i] = model_r.bbeta_sn1[i] * ttheta[idx]; idx += 1
        model_p.bbeta_sn2[i] = model_r.bbeta_sn2[i] * ttheta[idx]; idx += 1

        ## Consumption and growth (I * 4)
        model_p.aalpha_1[i] = r2p(p2r(model_r.aalpha_1[i] * model_r.aalpha_2[i]) * ttheta[idx]); idx += 1 # [0,1]
        model_p.aalpha_2[i] = torch.ones_like(model_p.aalpha_2[i]) # Un-identifiable
        model_p.ggamma_1[i] = r2p(p2r(model_r.ggamma_1[i]) * ttheta[idx]); idx += 1 # [0,1]
        model_p.ggamma_2[i] = model_r.ggamma_2[i] * ttheta[idx]; idx += 1

        ## Reproduction (I * 5)
        model_p.bbeta_b0[i] = r2p(p2r(model_r.bbeta_b0[i]) * ttheta[idx]); idx += 1 # [0,1]
        model_p.bbeta_b1[i] = model_r.bbeta_b1[i] * ttheta[idx]; idx += 1
        model_p.bbeta_b2[i] = model_r.bbeta_b2[i] * ttheta[idx]; idx += 1        
        model_p.nnu_o[i] = model_r.nnu_o[i] * ttheta[idx]; idx += 1
        model_p.mmu_o[i] = model_r.mmu_o[i] * ttheta[idx]; idx += 1

        ## Metabolic costs (I * 1)
        model_p.ddelta[i] = model_r.ddelta[i] * ttheta[idx]; idx += 1
    """
    
    ## Simulate model
    predictions = model_p.simulate(t_max=30, Dt=Dt0_p, Bt=Bt0_p, Pt=Pt0_p, Gt=Gt0_p, NNt=NNt0_p, NNNzt=NNNzt0_p, zzmax=zzmax, zzz_len=zzz_len)

    return predictions

#
###
```

```python
####################
## LAPLACE PRIORS ##
####################

def log_prior(ttheta, sigma):
    return -1/sigma * torch.sum(torch.abs(ttheta - 1))

print(log_prior(ttheta*1.3, .2))
    
#
###
```

```python
################
## LIKELIHOOD ##
################

def log_likelihood(ttheta):

    ## Predict
    predictions = wrapper(ttheta)
    
    ## Format
    time_series_tensor_predicted = format_predicted_time_series(predictions)

    ## Check for nans
    if torch.sum(torch.isnan(time_series_tensor_predicted))>1:
        log_lik = -float('inf')
    else:        
        if torch.prod(time_series_tensor_predicted[-1,3:]>1.0) == 0.0:

            log_lik = -float('inf')        

        else:
            
            ## Standardise
            time_series_tensor_predicted_standardised = apply_standardization(format_predicted_time_series(predictions), MEAN, STD)
            # time_series_tensor_standardised = apply_standardization(time_series_tensor, MEAN, STD)
            
            ## Compute residuals
            res = time_series_tensor_predicted_standardised - time_series_tensor_standardised
    
            ## 
            log_lik = - torch.nanmean(res[:,1:]**2)

    ## End
    return log_lik

#
###
```

```python
###############
## POSTERIOR ##
###############

def log_posterior(ttheta, sigma=0.2):
    log_lik = log_likelihood(ttheta)
    log_pri = log_prior(ttheta, sigma)
    log_pos = log_lik + log_pri
    return log_pos, log_lik, log_pri

t0 = time.time()
ttheta_0 = torch.ones(num_initial_conditions + num_parameters, dtype=torch.float64)
print(log_posterior(ttheta_0))
tf = time.time()
print(f"time: {tf-t0:2f}s")
    
#
###
```

```python
###############
## RUN CHAIN ##
###############

## Import module
from f_demc import de_mcmc_single_chain

def d_target(ttheta):
    log_pos, _, _ = log_posterior(ttheta, sigma=0.5)
    return log_pos

## Scale parameter vector
I = 6
num_parameters = 0
# num_parameters = I*2 + I*I*4 + I*3 + I*4 + I*5 + I*1
num_initial_conditions = 4 + I
ttheta_0 = torch.ones(num_initial_conditions + num_parameters, dtype=torch.float64)

## Run MCMC
t0 = time.time()
results = de_mcmc_single_chain(
    d_target=d_target,
    theta_0=ttheta_0,
    epsilon=1e-02,
    n_it=10000,
    verbose=True  # set True for intermediate output
)
tf = time.time()
print(f"time: {tf-t0:2f}s")

## Extract chain
chain = results['chain']
samples = chain[:, 1:].numpy()  # remove log-posterior column

#
###
```
