```python
##############
## INITIATE ##
##############

## Imports
import torch
import numpy as np
import matplotlib.pyplot as plt
import time
import seaborn as sns
import yaml

## Import modules
from f_YEM_V1_14_2025_10_10 import pdf
from f_YEM_V1_14_2025_10_10 import f_normal
from f_YEM_V1_14_2025_10_10 import YellowstoneEcosystemModel
from f_YEM_V1_14_2025_10_10 import plot_model

## Set the device to MPS if available
device = torch.device("mps" if torch.backends.mps.is_built() else "cpu")
device = "cpu" # Force cpu for now
print(device)

## Figure parameters
plt.rcParams['figure.dpi'] = 100
FIGSIZE = (6,4)
custom_params = {"axes.spines.right": False, "axes.spines.top": False, 'axes.facecolor':'whitesmoke'}
sns.set_theme(style="ticks", rc=custom_params)

## Font
PAPER_FORMATTING = False
if PAPER_FORMATTING == True:
    plt.rcParams['font.family'] = ['sans-serif']
    plt.rcParams['font.sans-serif'] = ['Times New Roman']
    plt.rcParams['font.size'] = 14
    plt.rcParams['mathtext.fontset'] = 'custom'
    plt.rcParams['mathtext.rm'] = 'Times New Roman'
    plt.rcParams['mathtext.it'] = 'Times New Roman:italic'
    plt.rcParams['mathtext.bf'] = 'Times New Roman:bold'

#
###
```

```python
####################
## INITIATE MODEL ##
####################

## Model
model = YellowstoneEcosystemModel('parameters-apriori-V1-14-2025-10-10.yaml')

## Load the YAML file
with open('initial-conditions-apriori-V1-14-2025-10-10.yaml', 'r') as file:
    initial_conditions = yaml.safe_load(file)

## Convert to tensors
Dt0 = torch.tensor(initial_conditions['Dt0'], dtype=torch.float64)
Bt0 = torch.tensor(initial_conditions['Bt0'], dtype=torch.float64)
Pt0 = torch.tensor(initial_conditions['Pt0'], dtype=torch.float64)
Gt0 = torch.tensor(initial_conditions['Gt0'], dtype=torch.float64)
NNt0 = torch.tensor(initial_conditions['NNt0'], dtype=torch.float64)
NNNzt0 = torch.tensor(initial_conditions['NNNzt0'], dtype=torch.float64)

## Load settings
with open('settings-apriori-V1-14-2025-10-10.yaml', 'r') as file:
    settings = yaml.safe_load(file)

## Convert to tensor
zzmax = torch.tensor(settings['zzmax'], dtype=torch.float64)
zzz_len = settings['zzz_len']

#
###
```

```python
###############
## RUN MODEL ##
###############

## Initiate parameters
t_max = 500

## Period 1
t0 = time.time()
predictions = model.simulate(t_max=t_max,
                             Dt = Dt0,
                             Bt = Bt0,
                             Pt = Pt0,
                             Gt = Gt0,
                             NNt = NNt0,
                             NNNzt = NNNzt0,
                             zzmax = zzmax,
                             zzz_len = zzz_len
                            )
zzz, tt, DDt, BBt, PPt, GGt, NNNt, NNNNzt, ttotal_biomass = predictions
tf = time.time()
print(f"{tf-t0:.2f}s")

## Visualise
plot_model(predictions)

#
###
```

```python
#########################################
## FUNCTIONS ANALYSIS OF CHAIN RESULTS ##
#########################################

## Note: To convert to an import as shared between C1 and C2.

def get_map(chain):
    """
    """
    idx = torch.argmax(chain[:,0])
    ttheta_map = chain[idx,1:]
    return ttheta_map
     
#
###
```

```python
#############################
## COMPARE EACH PARAMETERS ##
#############################

import torch
import numpy as np

def format_tensor(tensor, precision=4, scientific=False):
    """Detach, convert, and format a PyTorch tensor."""
    if torch.is_tensor(tensor):
        tensor = tensor.detach().cpu().numpy()
    if isinstance(tensor, (list, tuple, np.ndarray)):
        fmt = f"{{:.{precision}e}}" if scientific else f"{{:.{precision}f}}"
        return "[" + ", ".join(fmt.format(x) for x in tensor) + "]"
    else:
        fmt = f"{{:.{precision}e}}" if scientific else f"{{:.{precision}f}}"
        return fmt.format(tensor)

def print_parameter_comparison(name, val_r, val_p, indent=2, precision=4, scientific=False):
    pad = ' ' * indent
    print(f"{name}:")
    print(f"{pad}model_r: {format_tensor(val_r, precision, scientific)}")
    print(f"{pad}model_p: {format_tensor(val_p, precision, scientific)}")
    print("")

print("\n====================")
print("== ECOSYSTEM ==")
print("====================")
print(f"phi_BD: {model_r.phi_BD:.4f} {model_p.phi_BD:.4f}")
print(f"phi_DB: {model_r.phi_DB:.4f} {model_p.phi_DB:.4f}")
print(f"rho_B: {model_r.rho_B:.4f} {model_p.rho_B:.4f}")
print(f"phi_DP: {model_r.phi_DP:.4f} {model_p.phi_DP:.4f}")
print(f"phi_GD: {model_r.phi_GD:.4f} {model_p.phi_GD:.4f}")
print(f"phi_PG: {model_r.phi_PG:.4f} {model_p.phi_PG:.4f}")
print(f"rho_G: {model_r.rho_G:.4f} {model_p.rho_G:.4f}")
print(f"gamma_3_G: {model_r.gamma_3_G:.4f} {model_p.gamma_3_G:.4f}")

print("====================")
print("== NUTRIENTS & HERBIVORY ==")
print("====================")
print_parameter_comparison("bbeta_dh0_P", model_r.bbeta_dh0_P, model_p.bbeta_dh0_P)
print_parameter_comparison("bbeta_dh0_G", model_r.bbeta_dh0_G, model_p.bbeta_dh0_G)

print("====================")
print("== PREDATION ==")
for i in range(6):
    print(f"-- Species {i+1} --")
    print_parameter_comparison(f"bbbeta_dp0[{i}]", model_r.bbbeta_dp0[i], model_p.bbbeta_dp0[i])
    print_parameter_comparison(f"bbbeta_dp1[{i}]", model_r.bbbeta_dp1[i], model_p.bbbeta_dp1[i])
    print_parameter_comparison(f"bbbeta_dp2[{i}]", model_r.bbbeta_dp2[i], model_p.bbbeta_dp2[i])

print("====================")
print("== SCAVENGING ==")
for i in range(6):
    print(f"-- Species {i+1} --")
    print_parameter_comparison(f"lllambda[{i}]", model_r.lllambda[i], model_p.lllambda[i])
    # print_parameter_comparison(f"bbbeta_dn0[{i}]", model_r.bbbeta_dn0[i], model_p.bbbeta_dn0[i])

print("====================")
print("== SURVIVAL ==")
print_parameter_comparison("bbeta_sn0", model_r.bbeta_sn0, model_p.bbeta_sn0)
print_parameter_comparison("bbeta_sn1", model_r.bbeta_sn1, model_p.bbeta_sn1)
print_parameter_comparison("bbeta_sn2", model_r.bbeta_sn2, model_p.bbeta_sn2)

print("====================")
print("== REPRODUCTION ==")
print_parameter_comparison("bbeta_b0", model_r.bbeta_b0, model_p.bbeta_b0)
print_parameter_comparison("bbeta_b1", model_r.bbeta_b1, model_p.bbeta_b1)
print_parameter_comparison("bbeta_b2", model_r.bbeta_b2, model_p.bbeta_b2)
print_parameter_comparison("nnu_o", model_r.nnu_o, model_p.nnu_o)
print_parameter_comparison("mmu_o", model_r.mmu_o, model_p.mmu_o)

print("====================")
print("== ASSIMILATION ==")
print_parameter_comparison("aalpha_1", model_r.aalpha_1, model_p.aalpha_1)
print_parameter_comparison("aalpha_2", model_r.aalpha_2, model_p.aalpha_2)

print("====================")
print("== CONSUMPTION ==")
print_parameter_comparison("ggamma_1", model_r.ggamma_1, model_p.ggamma_1)
print_parameter_comparison("ggamma_2", model_r.ggamma_2, model_p.ggamma_2)

print("====================")
print("== MAINTENANCE ==")
print_parameter_comparison("ddelta", model_r.ddelta, model_p.ddelta)

#
##
```
